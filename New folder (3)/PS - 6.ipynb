```python
# ============================================================
# Continuous Bag of Words (CBOW) Model Implementation
# ============================================================

# a. Import required libraries
import tensorflow as tf
from tensorflow.keras import models, layers, optimizers
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import re
from sklearn.manifold import TSNE
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

# Download NLTK data (run once)
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')

try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('stopwords')

print(f"TensorFlow version: {tf.__version__}")

# For reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# ============================================================
# a. Data Preparation
# ============================================================

print("=" * 60)
print("STAGE A: DATA PREPARATION")
print("=" * 60)

def load_and_preprocess_text_data():
    """
    Load and preprocess textual data for CBOW model
    """
    # Sample textual document (you can replace this with any text dataset)
    sample_text = """
    Natural language processing is a subfield of linguistics and computer science. 
    It focuses on the interactions between computers and human language. 
    Machine learning models can understand and process natural language. 
    Deep learning has revolutionized natural language processing in recent years. 
    Word embeddings represent words as vectors in continuous space. 
    Continuous Bag of Words is a popular word embedding technique. 
    Neural networks learn contextual relationships between words. 
    Language models predict the probability of word sequences. 
    Text preprocessing is essential for natural language processing tasks. 
    Tokenization splits text into individual words or tokens.
    """
    
    # Alternative: Load from file (uncomment to use your own file)
    # with open('your_text_file.txt', 'r', encoding='utf-8') as file:
    #     sample_text = file.read()
    
    print("ğŸ“„ SAMPLE TEXT DOCUMENT:")
    print(sample_text[:500] + "..." if len(sample_text) > 500 else sample_text)
    
    return sample_text

def preprocess_text(text):
    """
    Preprocess the text: lowercase, remove punctuation, tokenize
    """
    # Convert to lowercase
    text = text.lower()
    
    # Remove punctuation and numbers
    text = re.sub(r'[^\w\s]', ' ', text)
    text = re.sub(r'\d+', '', text)
    
    # Tokenize
    tokens = word_tokenize(text)
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words and len(token) > 1]
    
    return tokens

# Load and preprocess data
text_data = load_and_preprocess_text_data()
tokens = preprocess_text(text_data)

print(f"\nğŸ“Š TEXT STATISTICS:")
print(f"Total tokens: {len(tokens)}")
print(f"Unique tokens: {len(set(tokens))}")
print(f"First 20 tokens: {tokens[:20]}")

# Build vocabulary
word_freq = Counter(tokens)
vocab = sorted(word_freq, key=word_freq.get, reverse=True)
word_to_index = {word: idx for idx, word in enumerate(vocab)}
index_to_word = {idx: word for idx, word in enumerate(vocab)}

vocab_size = len(vocab)
print(f"\nğŸ“š VOCABULARY:")
print(f"Vocabulary size: {vocab_size}")
print(f"Most common words: {word_freq.most_common(10)}")

# ============================================================
# b. Generate Training Data
# ============================================================

print("\n" + "=" * 60)
print("STAGE B: GENERATE TRAINING DATA")
print("=" * 60)

def generate_cbow_data(tokens, word_to_index, window_size=2):
    """
    Generate training data for CBOW model
    """
    X = []  # Context words
    y = []  # Target words
    
    for i in range(window_size, len(tokens) - window_size):
        # Get context words (surrounding words)
        context = []
        for j in range(-window_size, window_size + 1):
            if j != 0:  # Skip the target word itself
                context.append(tokens[i + j])
        
        # Get target word
        target = tokens[i]
        
        # Convert to indices
        context_indices = [word_to_index[word] for word in context]
        target_index = word_to_index[target]
        
        X.append(context_indices)
        y.append(target_index)
    
    return np.array(X), np.array(y)

# Generate training data
window_size = 2
X, y = generate_cbow_data(tokens, word_to_index, window_size)

print(f"ğŸ“Š TRAINING DATA SHAPE:")
print(f"Context words (X): {X.shape}")
print(f"Target words (y): {y.shape}")
print(f"Window size: {window_size}")
print(f"Context words per sample: {X.shape[1]}")

# Display sample training data
print(f"\nğŸ” SAMPLE TRAINING DATA:")
for i in range(5):
    context_words = [index_to_word[idx] for idx in X[i]]
    target_word = index_to_word[y[i]]
    print(f"Context: {context_words} -> Target: {target_word}")

# ============================================================
# c. Train Model
# ============================================================

print("\n" + "=" * 60)
print("STAGE C: TRAIN CBOW MODEL")
print("=" * 60)

def build_cbow_model(vocab_size, embedding_dim=50, window_size=2):
    """
    Build CBOW model architecture
    """
    model = models.Sequential([
        # Embedding layer for context words
        layers.Embedding(input_dim=vocab_size, 
                        output_dim=embedding_dim, 
                        input_length=window_size * 2,
                        name='embedding'),
        
        # Average the embeddings of context words
        layers.GlobalAveragePooling1D(name='average_pooling'),
        
        # Dense layer with ReLU activation
        layers.Dense(128, activation='relu', name='hidden_layer'),
        layers.Dropout(0.3),
        
        # Output layer with softmax for word prediction
        layers.Dense(vocab_size, activation='softmax', name='output')
    ])
    
    return model

# Model parameters
embedding_dim = 50
window_size = 2

# Build model
cbow_model = build_cbow_model(vocab_size, embedding_dim, window_size)

print("ğŸ”§ CBOW MODEL ARCHITECTURE:")
cbow_model.summary()

# Compile the model
cbow_model.compile(
    optimizer=optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("\nâœ… MODEL COMPILED:")
print(f"Optimizer: Adam (lr=0.001)")
print(f"Loss: Sparse Categorical Crossentropy")
print(f"Metrics: Accuracy")

# Train the model
print("\nğŸš€ TRAINING CBOW MODEL...")
history = cbow_model.fit(
    X, y,
    epochs=100,
    batch_size=32,
    validation_split=0.2,
    verbose=1,
    shuffle=True
)

print("\nâœ… Training completed!")

# ============================================================
# d. Output
# ============================================================

print("\n" + "=" * 60)
print("STAGE D: OUTPUT AND EVALUATION")
print("=" * 60)

# Training history visualization
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss', linewidth=2)
plt.plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
plt.title('CBOW Model Training Loss', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy', linewidth=2)
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', linewidth=2)
plt.title('CBOW Model Training Accuracy', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Extract word embeddings
embedding_layer = cbow_model.get_layer('embedding')
weights = embedding_layer.get_weights()[0]
print(f"\nğŸ“Š WORD EMBEDDINGS:")
print(f"Embedding matrix shape: {weights.shape}")

# Visualize word embeddings using t-SNE
def visualize_embeddings(weights, words, num_words=20):
    """
    Visualize word embeddings using t-SNE
    """
    # Get top N words
    top_words = words[:num_words]
    top_indices = [word_to_index[word] for word in top_words]
    top_embeddings = weights[top_indices]
    
    # Apply t-SNE
    tsne = TSNE(n_components=2, random_state=42, perplexity=min(5, num_words-1))
    embeddings_2d = tsne.fit_transform(top_embeddings)
    
    # Plot
    plt.figure(figsize=(12, 8))
    plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1], alpha=0.7)
    
    for i, word in enumerate(top_words):
        plt.annotate(word, (embeddings_2d[i, 0], embeddings_2d[i, 1]), 
                    fontsize=12, alpha=0.8)
    
    plt.title('Word Embeddings Visualization (t-SNE)', fontsize=16, fontweight='bold')
    plt.xlabel('t-SNE Component 1')
    plt.ylabel('t-SNE Component 2')
    plt.grid(True, alpha=0.3)
    plt.show()

# Visualize top words
top_words = [word for word, _ in word_freq.most_common(20)]
visualize_embeddings(weights, top_words)

# Function to find similar words
def find_similar_words(word, weights, word_to_index, index_to_word, top_n=5):
    """
    Find similar words based on cosine similarity
    """
    if word not in word_to_index:
        return f"Word '{word}' not in vocabulary"
    
    # Get word vector
    word_idx = word_to_index[word]
    word_vec = weights[word_idx]
    
    # Calculate cosine similarities
    similarities = {}
    for idx, vec in enumerate(weights):
        if idx != word_idx:
            cosine_sim = np.dot(word_vec, vec) / (np.linalg.norm(word_vec) * np.linalg.norm(vec))
            similarities[index_to_word[idx]] = cosine_sim
    
    # Get top N similar words
    similar_words = sorted(similarities.items(), key=lambda x: x[1], reverse=True)[:top_n]
    
    return similar_words

# Test similar words
print("\nğŸ” SIMILAR WORD ANALYSIS:")
test_words = ['natural', 'language', 'learning', 'processing']
for word in test_words:
    if word in word_to_index:
        similar = find_similar_words(word, weights, word_to_index, index_to_word)
        print(f"Words similar to '{word}': {similar}")

# Word prediction function
def predict_target_word(context_words, model, word_to_index, index_to_word, top_k=3):
    """
    Predict target word given context words
    """
    # Convert context words to indices
    context_indices = [word_to_index[word] for word in context_words if word in word_to_index]
    
    if len(context_indices) != window_size * 2:
        return "Invalid context words"
    
    # Reshape for prediction
    context_array = np.array([context_indices])
    
    # Make prediction
    predictions = model.predict(context_array, verbose=0)[0]
    
    # Get top K predictions
    top_indices = np.argsort(predictions)[-top_k:][::-1]
    top_words = [(index_to_word[idx], predictions[idx]) for idx in top_indices]
    
    return top_words

# Test word prediction
print(f"\nğŸ¯ WORD PREDICTION EXAMPLES:")
test_contexts = [
    ['natural', 'language', 'revolutionized', 'deep'],
    ['word', 'embeddings', 'represent', 'words'],
    ['machine', 'learning', 'models', 'understand']
]

for context in test_contexts:
    predictions = predict_target_word(context, cbow_model, word_to_index, index_to_word)
    print(f"Context: {context} -> Predictions: {predictions}")

# Save word embeddings
def save_embeddings(weights, word_to_index, filename='cbow_embeddings.txt'):
    """
    Save word embeddings to file
    """
    with open(filename, 'w', encoding='utf-8') as f:
        f.write(f"{vocab_size} {embedding_dim}\n")
        for word, idx in word_to_index.items():
            vec = ' '.join(map(str, weights[idx]))
            f.write(f"{word} {vec}\n")
    print(f"âœ… Word embeddings saved to {filename}")

save_embeddings(weights, word_to_index)

# Final model evaluation
print(f"\nğŸ“ˆ FINAL MODEL EVALUATION:")
final_loss, final_accuracy = cbow_model.evaluate(X, y, verbose=0)
print(f"Final Training Loss: {final_loss:.4f}")
print(f"Final Training Accuracy: {final_accuracy:.4f}")

# Display vocabulary statistics
print(f"\nğŸ“š VOCABULARY STATISTICS:")
print(f"Total words in vocabulary: {vocab_size}")
print(f"Embedding dimension: {embedding_dim}")
print(f"Window size: {window_size}")
print(f"Training samples: {len(X)}")

# Show some learned relationships
print(f"\nğŸ”— LEARNED WORD RELATIONSHIPS:")
relationship_examples = [
    (['natural', 'language'], 'processing'),
    (['word', 'embeddings'], 'represent'),
    (['machine', 'learning'], 'models')
]

for context, expected in relationship_examples:
    if all(word in word_to_index for word in context):
        prediction = predict_target_word(context, cbow_model, word_to_index, index_to_word, top_k=1)
        print(f"Context: {context} | Expected: '{expected}' | Predicted: '{prediction[0][0]}'")

print("\n" + "=" * 70)
print("CBOW MODEL IMPLEMENTATION COMPLETE!")
print("=" * 70)
```

```python
# ============================================================
# Continuous Bag of Words (CBOW) Model Implementation
# Enhanced Version with File Loading Capabilities
# ============================================================

# a. Import required libraries
import tensorflow as tf
from tensorflow.keras import models, layers, optimizers
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import re
import os
from sklearn.manifold import TSNE
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string
import requests
import zipfile
import gzip

# Download NLTK data (run once)
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')

try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('stopwords')

print(f"TensorFlow version: {tf.__version__}")

# For reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# ============================================================
# a. Data Preparation with Multiple Loading Options
# ============================================================

print("=" * 60)
print("STAGE A: DATA PREPARATION WITH FILE LOADING")
print("=" * 60)

def load_text_from_file(file_path):
    """
    Load text from various file formats
    """
    supported_extensions = ['.txt', '.csv', '.json', '.xml']
    file_ext = os.path.splitext(file_path)[1].lower()
    
    try:
        if file_ext == '.txt':
            with open(file_path, 'r', encoding='utf-8') as file:
                return file.read()
        
        elif file_path.endswith('.gz'):
            with gzip.open(file_path, 'rt', encoding='utf-8') as file:
                return file.read()
                
        else:
            print(f"âš ï¸  Unsupported file format: {file_ext}")
            print("Supported formats: .txt, .gz")
            return None
            
    except FileNotFoundError:
        print(f"âŒ File not found: {file_path}")
        return None
    except Exception as e:
        print(f"âŒ Error reading file: {e}")
        return None

def load_sample_dataset(dataset_name='nltk'):
    """
    Load sample datasets for demonstration
    """
    if dataset_name == 'nltk':
        try:
            from nltk.corpus import gutenberg
            nltk.download('gutenberg')
            text = gutenberg.raw('shakespeare-hamlet.txt')
            print("âœ… Loaded Shakespeare's Hamlet from NLTK")
            return text
        except:
            print("âš ï¸  NLTK Gutenberg corpus not available")
    
    elif dataset_name == 'simple':
        return """
        Machine learning is a method of data analysis that automates analytical model building. 
        It is a branch of artificial intelligence based on the idea that systems can learn from data. 
        Deep learning is a subset of machine learning that uses neural networks with many layers. 
        Natural language processing enables computers to understand human language. 
        Computer vision allows machines to interpret and understand visual world.
        """
    
    return None

def load_text_from_url(url):
    """
    Load text from a URL
    """
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        return response.text
    except Exception as e:
        print(f"âŒ Error loading from URL: {e}")
        return None

def get_available_files(directory='.'):
    """
    Get available text files in current directory
    """
    text_files = []
    for file in os.listdir(directory):
        if file.endswith(('.txt', '.csv', '.json', '.xml', '.gz')):
            text_files.append(file)
    return text_files

def load_and_preprocess_text_data():
    """
    Load and preprocess textual data from multiple sources
    """
    print("\nğŸ“ AVAILABLE DATA SOURCES:")
    print("1. Local file")
    print("2. Sample dataset")
    print("3. URL")
    print("4. Current directory files")
    
    choice = input("\nChoose data source (1-4): ").strip()
    
    if choice == '1':
        file_path = input("Enter file path: ").strip()
        text_data = load_text_from_file(file_path)
        
    elif choice == '2':
        print("\nAvailable sample datasets:")
        print("1. NLTK Gutenberg (Shakespeare)")
        print("2. Simple ML text")
        sample_choice = input("Choose sample dataset (1-2): ").strip()
        
        if sample_choice == '1':
            text_data = load_sample_dataset('nltk')
        else:
            text_data = load_sample_dataset('simple')
            
    elif choice == '3':
        url = input("Enter URL: ").strip()
        text_data = load_text_from_url(url)
        
    elif choice == '4':
        available_files = get_available_files()
        if available_files:
            print("\nAvailable files:")
            for i, file in enumerate(available_files, 1):
                print(f"{i}. {file}")
            file_choice = input("Choose file number: ").strip()
            try:
                selected_file = available_files[int(file_choice) - 1]
                text_data = load_text_from_file(selected_file)
            except (ValueError, IndexError):
                print("âŒ Invalid selection")
                text_data = None
        else:
            print("âŒ No text files found in current directory")
            text_data = None
    else:
        print("âŒ Invalid choice")
        text_data = None
    
    # Fallback to default sample text
    if text_data is None:
        print("âš ï¸  Using default sample text")
        text_data = """
        Natural language processing is a subfield of linguistics and computer science. 
        It focuses on the interactions between computers and human language. 
        Machine learning models can understand and process natural language. 
        Deep learning has revolutionized natural language processing in recent years. 
        Word embeddings represent words as vectors in continuous space. 
        Continuous Bag of Words is a popular word embedding technique. 
        Neural networks learn contextual relationships between words. 
        Language models predict the probability of word sequences. 
        Text preprocessing is essential for natural language processing tasks. 
        Tokenization splits text into individual words or tokens.
        """
    
    return text_data

def preprocess_text(text, remove_stopwords=True, min_word_length=2):
    """
    Preprocess the text with configurable options
    """
    # Convert to lowercase
    text = text.lower()
    
    # Remove punctuation and numbers
    text = re.sub(r'[^\w\s]', ' ', text)
    text = re.sub(r'\d+', '', text)
    
    # Tokenize
    tokens = word_tokenize(text)
    
    if remove_stopwords:
        # Remove stopwords
        stop_words = set(stopwords.words('english'))
        tokens = [token for token in tokens if token not in stop_words and len(token) >= min_word_length]
    else:
        tokens = [token for token in tokens if len(token) >= min_word_length]
    
    return tokens

# Load and preprocess data
print("\nğŸ”„ LOADING TEXT DATA...")
text_data = load_and_preprocess_text_data()

print(f"\nğŸ“„ TEXT PREVIEW (first 500 characters):")
print(text_data[:500] + "..." if len(text_data) > 500 else text_data)

# Ask for preprocessing options
print("\nâš™ï¸  PREPROCESSING OPTIONS:")
remove_stopwords = input("Remove stopwords? (y/n): ").strip().lower() == 'y'
min_length = input("Minimum word length (default 2): ").strip()
min_word_length = int(min_length) if min_length.isdigit() else 2

tokens = preprocess_text(text_data, remove_stopwords, min_word_length)

print(f"\nğŸ“Š TEXT STATISTICS:")
print(f"Total tokens: {len(tokens):,}")
print(f"Unique tokens: {len(set(tokens)):,}")
print(f"Tokenized sample: {tokens[:20]}")

# Build vocabulary
word_freq = Counter(tokens)
vocab = sorted(word_freq, key=word_freq.get, reverse=True)
word_to_index = {word: idx for idx, word in enumerate(vocab)}
index_to_word = {idx: word for idx, word in enumerate(vocab)}

vocab_size = len(vocab)
print(f"\nğŸ“š VOCABULARY:")
print(f"Vocabulary size: {vocab_size}")
print(f"Most common words: {word_freq.most_common(15)}")

# ============================================================
# b. Generate Training Data
# ============================================================

print("\n" + "=" * 60)
print("STAGE B: GENERATE TRAINING DATA")
print("=" * 60)

def generate_cbow_data(tokens, word_to_index, window_size=2):
    """
    Generate training data for CBOW model
    """
    X = []  # Context words
    y = []  # Target words
    
    for i in range(window_size, len(tokens) - window_size):
        # Get context words (surrounding words)
        context = []
        for j in range(-window_size, window_size + 1):
            if j != 0:  # Skip the target word itself
                context.append(tokens[i + j])
        
        # Get target word
        target = tokens[i]
        
        # Convert to indices
        context_indices = [word_to_index[word] for word in context if word in word_to_index]
        
        # Only add if we have all context words
        if len(context_indices) == window_size * 2:
            target_index = word_to_index[target]
            X.append(context_indices)
            y.append(target_index)
    
    return np.array(X), np.array(y)

# Get window size from user
window_input = input("Enter window size (default 2): ").strip()
window_size = int(window_input) if window_input.isdigit() else 2

# Generate training data
X, y = generate_cbow_data(tokens, word_to_index, window_size)

print(f"ğŸ“Š TRAINING DATA SHAPE:")
print(f"Context words (X): {X.shape}")
print(f"Target words (y): {y.shape}")
print(f"Window size: {window_size}")
print(f"Context words per sample: {X.shape[1]}")
print(f"Total training samples: {len(X):,}")

# Display sample training data
print(f"\nğŸ” SAMPLE TRAINING DATA:")
for i in range(min(5, len(X))):
    context_words = [index_to_word[idx] for idx in X[i]]
    target_word = index_to_word[y[i]]
    print(f"Context: {context_words} -> Target: '{target_word}'")

# ============================================================
# c. Train Model
# ============================================================

print("\n" + "=" * 60)
print("STAGE C: TRAIN CBOW MODEL")
print("=" * 60)

def build_cbow_model(vocab_size, embedding_dim=50, window_size=2):
    """
    Build CBOW model architecture
    """
    model = models.Sequential([
        # Embedding layer for context words
        layers.Embedding(input_dim=vocab_size, 
                        output_dim=embedding_dim, 
                        input_length=window_size * 2,
                        name='embedding'),
        
        # Average the embeddings of context words
        layers.GlobalAveragePooling1D(name='average_pooling'),
        
        # Dense layer with ReLU activation
        layers.Dense(128, activation='relu', name='hidden_layer'),
        layers.Dropout(0.3),
        
        # Output layer with softmax for word prediction
        layers.Dense(vocab_size, activation='softmax', name='output')
    ])
    
    return model

# Get model parameters from user
embedding_input = input("Enter embedding dimension (default 50): ").strip()
embedding_dim = int(embedding_input) if embedding_input.isdigit() else 50

epochs_input = input("Enter number of epochs (default 100): ").strip()
epochs = int(epochs_input) if epochs_input.isdigit() else 100

batch_input = input("Enter batch size (default 32): ").strip()
batch_size = int(batch_input) if batch_input.isdigit() else 32

# Build model
cbow_model = build_cbow_model(vocab_size, embedding_dim, window_size)

print("ğŸ”§ CBOW MODEL ARCHITECTURE:")
cbow_model.summary()

# Compile the model
cbow_model.compile(
    optimizer=optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("\nâœ… MODEL COMPILED:")
print(f"Optimizer: Adam (lr=0.001)")
print(f"Loss: Sparse Categorical Crossentropy")
print(f"Metrics: Accuracy")
print(f"Embedding dimension: {embedding_dim}")
print(f"Epochs: {epochs}")
print(f"Batch size: {batch_size}")

# Train the model
print("\nğŸš€ TRAINING CBOW MODEL...")
history = cbow_model.fit(
    X, y,
    epochs=epochs,
    batch_size=batch_size,
    validation_split=0.2,
    verbose=1,
    shuffle=True
)

print("\nâœ… Training completed!")

# ============================================================
# d. Output and Results
# ============================================================

print("\n" + "=" * 60)
print("STAGE D: OUTPUT AND EVALUATION")
print("=" * 60)

# Training history visualization
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss', linewidth=2)
plt.plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
plt.title('CBOW Model Training Loss', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy', linewidth=2)
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', linewidth=2)
plt.title('CBOW Model Training Accuracy', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Extract word embeddings
embedding_layer = cbow_model.get_layer('embedding')
weights = embedding_layer.get_weights()[0]
print(f"\nğŸ“Š WORD EMBEDDINGS:")
print(f"Embedding matrix shape: {weights.shape}")

# Enhanced visualization function
def visualize_embeddings(weights, word_freq, num_words=25):
    """
    Visualize word embeddings using t-SNE
    """
    # Get top N words
    top_words = [word for word, _ in word_freq.most_common(num_words)]
    top_indices = [word_to_index[word] for word in top_words]
    top_embeddings = weights[top_indices]
    
    # Apply t-SNE
    tsne = TSNE(n_components=2, random_state=42, perplexity=min(10, num_words-1))
    embeddings_2d = tsne.fit_transform(top_embeddings)
    
    # Plot
    plt.figure(figsize=(15, 10))
    plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1], alpha=0.7, s=100)
    
    for i, word in enumerate(top_words):
        plt.annotate(word, (embeddings_2d[i, 0], embeddings_2d[i, 1]), 
                    fontsize=12, alpha=0.8,
                    bbox=dict(boxstyle="round,pad=0.3", fc="yellow", alpha=0.3))
    
    plt.title(f'Word Embeddings Visualization (Top {num_words} Words)', 
              fontsize=16, fontweight='bold')
    plt.xlabel('t-SNE Component 1')
    plt.ylabel('t-SNE Component 2')
    plt.grid(True, alpha=0.3)
    plt.show()

# Visualize top words
visualize_embeddings(weights, word_freq, 25)

# Enhanced similar words function
def find_similar_words(word, weights, word_to_index, index_to_word, top_n=8):
    """
    Find similar words based on cosine similarity
    """
    if word not in word_to_index:
        return f"Word '{word}' not in vocabulary"
    
    # Get word vector
    word_idx = word_to_index[word]
    word_vec = weights[word_idx]
    
    # Calculate cosine similarities
    similarities = {}
    for idx, vec in enumerate(weights):
        if idx != word_idx:
            cosine_sim = np.dot(word_vec, vec) / (np.linalg.norm(word_vec) * np.linalg.norm(vec))
            similarities[index_to_word[idx]] = cosine_sim
    
    # Get top N similar words
    similar_words = sorted(similarities.items(), key=lambda x: x[1], reverse=True)[:top_n]
    
    return similar_words

# Interactive similar words analysis
print("\nğŸ” SIMILAR WORD ANALYSIS:")
print("Enter words to find similar words (press Enter with empty input to stop):")

while True:
    word = input("\nEnter a word: ").strip().lower()
    if not word:
        break
    if word in word_to_index:
        similar = find_similar_words(word, weights, word_to_index, index_to_word, 8)
        print(f"Words similar to '{word}':")
        for similar_word, score in similar:
            print(f"  {similar_word}: {score:.4f}")
    else:
        print(f"âŒ Word '{word}' not in vocabulary")
        print(f"Similar words in vocabulary: {[w for w in vocab if w.startswith(word[:3])][:5]}")

# Enhanced word prediction
def predict_target_word(context_words, model, word_to_index, index_to_word, top_k=5):
    """
    Predict target word given context words
    """
    # Convert context words to indices
    context_indices = [word_to_index[word] for word in context_words if word in word_to_index]
    
    if len(context_indices) != window_size * 2:
        return f"Need exactly {window_size * 2} context words"
    
    # Reshape for prediction
    context_array = np.array([context_indices])
    
    # Make prediction
    predictions = model.predict(context_array, verbose=0)[0]
    
    # Get top K predictions
    top_indices = np.argsort(predictions)[-top_k:][::-1]
    top_words = [(index_to_word[idx], predictions[idx]) for idx in top_indices]
    
    return top_words

# Interactive word prediction
print(f"\nğŸ¯ INTERACTIVE WORD PREDICTION:")
print(f"Enter {window_size * 2} context words (press Enter with empty input to stop):")

while True:
    context_input = input(f"\nEnter {window_size * 2} context words (space-separated): ").strip()
    if not context_input:
        break
    
    context_words = context_input.split()
    if len(context_words) != window_size * 2:
        print(f"âŒ Please enter exactly {window_size * 2} words")
        continue
    
    predictions = predict_target_word(context_words, cbow_model, word_to_index, index_to_word, 5)
    print(f"Context: {context_words}")
    print("Predictions:")
    for word, prob in predictions:
        print(f"  {word}: {prob:.4f}")

# Save results
def save_results(weights, word_to_index, word_freq, filename_prefix='cbow'):
    """
    Save all results to files
    """
    # Save word embeddings
    embedding_file = f"{filename_prefix}_embeddings.txt"
    with open(embedding_file, 'w', encoding='utf-8') as f:
        f.write(f"{vocab_size} {embedding_dim}\n")
        for word, idx in word_to_index.items():
            vec = ' '.join(map(str, weights[idx]))
            f.write(f"{word} {vec}\n")
    
    # Save vocabulary with frequencies
    vocab_file = f"{filename_prefix}_vocabulary.txt"
    with open(vocab_file, 'w', encoding='utf-8') as f:
        for word, freq in word_freq.most_common():
            f.write(f"{word}\t{freq}\n")
    
    # Save model summary
    summary_file = f"{filename_prefix}_summary.txt"
    with open(summary_file, 'w', encoding='utf-8') as f:
        f.write("CBOW Model Training Summary\n")
        f.write("=" * 40 + "\n")
        f.write(f"Vocabulary size: {vocab_size}\n")
        f.write(f"Embedding dimension: {embedding_dim}\n")
        f.write(f"Window size: {window_size}\n")
        f.write(f"Training samples: {len(X)}\n")
        f.write(f"Final training accuracy: {history.history['accuracy'][-1]:.4f}\n")
        f.write(f"Final validation accuracy: {history.history['val_accuracy'][-1]:.4f}\n")
    
    print(f"âœ… Results saved:")
    print(f"   - Embeddings: {embedding_file}")
    print(f"   - Vocabulary: {vocab_file}")
    print(f"   - Summary: {summary_file}")

save_results(weights, word_to_index, word_freq)

# Final comprehensive evaluation
print(f"\nğŸ“ˆ COMPREHENSIVE MODEL EVALUATION:")
final_loss, final_accuracy = cbow_model.evaluate(X, y, verbose=0)
print(f"Final Training Loss: {final_loss:.4f}")
print(f"Final Training Accuracy: {final_accuracy:.4f}")
print(f"Final Validation Accuracy: {history.history['val_accuracy'][-1]:.4f}")

print(f"\nğŸ“Š DATASET STATISTICS:")
print(f"Total words processed: {len(tokens):,}")
print(f"Vocabulary size: {vocab_size}")
print(f"Training samples: {len(X):,}")
print(f"Embedding dimension: {embedding_dim}")
print(f"Window size: {window_size}")

print(f"\nğŸ”— TOP WORD RELATIONSHIPS:")
top_words = [word for word, _ in word_freq.most_common(10)]
for i, word1 in enumerate(top_words[:5]):
    for word2 in top_words[i+1:5]:
        if word1 in word_to_index and word2 in word_to_index:
            vec1 = weights[word_to_index[word1]]
            vec2 = weights[word_to_index[word2]]
            similarity = np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))
            print(f"  {word1} - {word2}: {similarity:.4f}")

print("\n" + "=" * 70)
print("ğŸ‰ CBOW MODEL WITH FILE LOADING COMPLETE!")
print("=" * 70)
```
