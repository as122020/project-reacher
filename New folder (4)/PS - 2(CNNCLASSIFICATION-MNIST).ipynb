```python
# Implement the Image classification CNN model for classifying hand-written
# MNIST dataset by dividing the model into following 4 stages:
# a. Loading and preprocessing the image data
# b. Defining the model's architecture
# c. Training the model
# d. Estimating the model's performance
```

```python
# a. Import the necessary packages cnn heavy code will take 30 mins for 15 epochs 
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

print("=== MNIST CNN CLASSIFICATION ===")
print("Following 4-Stage Approach")
print("="*50)

# =============================================
# STAGE a: Loading and preprocessing the image data
# =============================================

def load_and_preprocess_data():
    """
    Stage a: Load and preprocess MNIST data from CSV files
    """
    print("\n" + "="*50)
    print("STAGE a: LOADING AND PREPROCESSING IMAGE DATA")
    print("="*50)
    
    # Load data from CSV files
    print("Loading training data from mnist_train.csv...")
    train_data = pd.read_csv('mnist_train.csv')
    print("Loading test data from mnist_test.csv...")
    test_data = pd.read_csv('mnist_test.csv')
    
    # Extract labels (first column) and images (remaining 784 columns)
    y_train = train_data.iloc[:, 0].values
    x_train = train_data.iloc[:, 1:].values
    
    y_test = test_data.iloc[:, 0].values
    x_test = test_data.iloc[:, 1:].values
    
    print(f"Raw training data shape: {x_train.shape}")
    print(f"Raw test data shape: {x_test.shape}")
    
    # Normalize pixel values to range [0, 1]
    x_train = x_train.astype('float32') / 255.0
    x_test = x_test.astype('float32') / 255.0
    
    # Reshape for CNN: (samples, 28, 28, 1) - Important for CNN!
    x_train = x_train.reshape(-1, 28, 28, 1)
    x_test = x_test.reshape(-1, 28, 28, 1)
    
    # Convert labels to categorical one-hot encoding
    y_train_categorical = keras.utils.to_categorical(y_train, 10)
    y_test_categorical = keras.utils.to_categorical(y_test, 10)
    
    print(f"\nAfter preprocessing:")
    print(f"Training images shape: {x_train.shape}")
    print(f"Training labels shape: {y_train_categorical.shape}")
    print(f"Test images shape: {x_test.shape}")
    print(f"Test labels shape: {y_test_categorical.shape}")
    print(f"Label distribution: {np.bincount(y_train)}")
    
    # Display sample images
    display_sample_images(x_train, y_train)
    
    return x_train, y_train_categorical, x_test, y_test_categorical, y_train, y_test

def display_sample_images(images, labels, num_samples=12):
    """
    Display sample images from the dataset
    """
    print("\nDisplaying sample images...")
    plt.figure(figsize=(12, 8))
    for i in range(num_samples):
        plt.subplot(3, 4, i + 1)
        plt.imshow(images[i].reshape(28, 28), cmap='gray')
        plt.title(f'Label: {labels[i]}', fontsize=12)
        plt.axis('off')
    plt.suptitle('Sample MNIST Images from Training Set', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()

# =============================================
# STAGE b: Defining the model's architecture
# =============================================

def create_cnn_model():
    """
    Stage b: Define CNN architecture for MNIST classification
    """
    print("\n" + "="*50)
    print("STAGE b: DEFINING CNN ARCHITECTURE")
    print("="*50)
    
    model = keras.Sequential([
        # First Convolutional Block
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        layers.BatchNormalization(),
        layers.Conv2D(32, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        
        # Second Convolutional Block
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        
        # Third Convolutional Block
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.25),
        
        # Flatten and Dense Layers
        layers.Flatten(),
        layers.Dense(256, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(128, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        
        # Output Layer
        layers.Dense(10, activation='softmax')
    ])
    
    # Compile the model
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    print("CNN Model Architecture:")
    model.summary()
    
    return model

# =============================================
# STAGE c: Training the model
# =============================================

def train_model(model, x_train, y_train, x_test, y_test):
    """
    Stage c: Train the CNN model
    """
    print("\n" + "="*50)
    print("STAGE c: TRAINING THE MODEL")
    print("="*50)
    
    # Define callbacks
    callbacks = [
        keras.callbacks.EarlyStopping(
            patience=5,
            restore_best_weights=True,
            monitor='val_accuracy'
        ),
        keras.callbacks.ReduceLROnPlateau(
            factor=0.5,
            patience=3,
            min_lr=1e-7
        ),
        keras.callbacks.ModelCheckpoint(
            'best_mnist_cnn_model.h5',
            save_best_only=True,
            monitor='val_accuracy'
        )
    ]
    
    print("Starting CNN training...")
    print(f"Training samples: {len(x_train)}")
    print(f"Validation samples: {len(x_test)}")
    print(f"Input shape: {x_train.shape[1:]}")
    
    # Train the model
    history = model.fit(
        x_train, y_train,
        batch_size=128,
        epochs=15,
        validation_data=(x_test, y_test),
        callbacks=callbacks,
        verbose=1
    )
    
    print("\nTraining completed successfully!")
    return history

# =============================================
# STAGE d: Estimating the model's performance
# =============================================

def evaluate_model_performance(model, history, x_test, y_test, y_test_original):
    """
    Stage d: Evaluate model performance comprehensively
    """
    print("\n" + "="*50)
    print("STAGE d: ESTIMATING MODEL PERFORMANCE")
    print("="*50)
    
    # 1. Basic Evaluation
    print("1. BASIC PERFORMANCE METRICS:")
    print("-" * 30)
    
    test_loss, test_accuracy = model.evaluate(x_test, y_test, verbose=0)
    print(f"Test Accuracy:  {test_accuracy:.4f}")
    print(f"Test Loss:      {test_loss:.4f}")
    
    # 2. Training History Visualization
    print("\n2. TRAINING HISTORY VISUALIZATION:")
    plot_training_history(history)
    
    # 3. Predictions and Analysis
    print("\n3. DETAILED PREDICTIONS ANALYSIS:")
    analyze_predictions(model, x_test, y_test, y_test_original)
    
    # 4. Performance Summary
    print("\n4. COMPREHENSIVE PERFORMANCE SUMMARY:")
    print("-" * 40)
    
    final_accuracy = history.history['val_accuracy'][-1]
    best_accuracy = max(history.history['val_accuracy'])
    
    print(f"Final Test Accuracy: {final_accuracy:.4f}")
    print(f"Best Test Accuracy:  {best_accuracy:.4f}")
    print(f"Training Samples:    {len(x_test) * 4:,}")  # 80% of 60,000
    print(f"Test Samples:        {len(x_test):,}")
    print(f"Model Parameters:    {model.count_params():,}")
    
    # Save final model
    model.save('mnist_cnn_final_model.h5')
    print(f"\n✓ Final model saved as 'mnist_cnn_final_model.h5'")
    
    return test_accuracy

def plot_training_history(history):
    """
    Plot training history
    """
    plt.figure(figsize=(15, 5))
    
    # Accuracy plot
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy', linewidth=2)
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy', linewidth=2)
    plt.title('Model Accuracy', fontsize=14, fontweight='bold')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # Loss plot
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss', linewidth=2)
    plt.plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
    plt.title('Model Loss', fontsize=14, fontweight='bold')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

def analyze_predictions(model, x_test, y_test, y_test_original):
    """
    Analyze model predictions in detail
    """
    # Get predictions
    predictions = model.predict(x_test, verbose=0)
    predicted_classes = np.argmax(predictions, axis=1)
    true_classes = y_test_original
    
    # Calculate overall accuracy
    overall_accuracy = np.mean(true_classes == predicted_classes)
    print(f"Overall Accuracy: {overall_accuracy:.4f}")
    
    # Sample predictions visualization
    plot_sample_predictions(x_test, true_classes, predicted_classes, predictions)
    
    # Confusion Matrix
    plot_confusion_matrix(true_classes, predicted_classes)
    
    # Classification Report
    print("\nClassification Report:")
    print("-" * 50)
    print(classification_report(true_classes, predicted_classes, digits=4))
    
    # Misclassified examples
    plot_misclassified_examples(x_test, true_classes, predicted_classes, predictions)

def plot_sample_predictions(images, true_labels, pred_labels, predictions, num_samples=12):
    """
    Plot sample predictions with images
    """
    plt.figure(figsize=(15, 10))
    for i in range(num_samples):
        plt.subplot(3, 4, i + 1)
        plt.imshow(images[i].reshape(28, 28), cmap='gray')
        
        true_label = true_labels[i]
        pred_label = pred_labels[i]
        confidence = np.max(predictions[i])
        
        color = 'green' if true_label == pred_label else 'red'
        plt.title(f'True: {true_label}, Pred: {pred_label}\nConf: {confidence:.3f}', 
                 color=color, fontsize=10)
        plt.axis('off')
    
    plt.suptitle('Sample Predictions on Test Set', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()

def plot_confusion_matrix(true_labels, pred_labels):
    """
    Plot confusion matrix
    """
    cm = confusion_matrix(true_labels, pred_labels)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title('Confusion Matrix', fontsize=14, fontweight='bold')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.tight_layout()
    plt.show()

def plot_misclassified_examples(images, true_labels, pred_labels, predictions, num_examples=12):
    """
    Plot misclassified examples
    """
    misclassified_indices = np.where(true_labels != pred_labels)[0]
    
    if len(misclassified_indices) > 0:
        print(f"\nNumber of misclassified samples: {len(misclassified_indices)}")
        print(f"Error rate: {len(misclassified_indices)/len(true_labels)*100:.2f}%")
        
        plt.figure(figsize=(15, 10))
        num_to_show = min(num_examples, len(misclassified_indices))
        
        for i, idx in enumerate(misclassified_indices[:num_to_show]):
            plt.subplot(3, 4, i + 1)
            plt.imshow(images[idx].reshape(28, 28), cmap='gray')
            
            true_label = true_labels[idx]
            pred_label = pred_labels[idx]
            confidence = np.max(predictions[idx])
            
            plt.title(f'True: {true_label}, Pred: {pred_label}\nConf: {confidence:.3f}', 
                     color='red', fontsize=10)
            plt.axis('off')
        
        plt.suptitle('Misclassified Examples', fontsize=14, fontweight='bold', color='red')
        plt.tight_layout()
        plt.show()
    else:
        print("\nNo misclassified samples! Perfect accuracy!")

# =============================================
# MAIN EXECUTION - 4 STAGES
# =============================================

def main():
    """
    Execute all 4 stages of CNN image classification
    """
    try:
        # Stage a: Loading and preprocessing
        x_train, y_train, x_test, y_test, y_train_orig, y_test_orig = load_and_preprocess_data()
        
        # Stage b: Model architecture
        model = create_cnn_model()
        
        # Stage c: Training
        history = train_model(model, x_train, y_train, x_test, y_test)
        
        # Stage d: Performance estimation
        final_accuracy = evaluate_model_performance(model, history, x_test, y_test, y_test_orig)
        
        # Final Summary
        print("\n" + "="*60)
        print("🎯 CNN CLASSIFICATION COMPLETED SUCCESSFULLY!")
        print("="*60)
        print(f"📊 Final Test Accuracy: {final_accuracy:.4f}")
        print(f"🏷️  Number of Classes: 10")
        print(f"🖼️  Input Shape: 28x28x1")
        print(f"🔧 Model Type: Convolutional Neural Network")
        print(f"✅ All 4 stages executed successfully!")
        print("="*60)
        
    except Exception as e:
        print(f"❌ Error in execution: {e}")
        import traceback
        traceback.print_exc()

# Run the complete pipeline
if __name__ == "__main__":
    main()
```

```python
# a. Import the necessary packages cnn lightweight code will take 1 min for entire training
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import os

# Set memory growth and limit GPU memory usage
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
        print(e)

# Force CPU usage for consistent performance in exams
os.environ['CUDA_VISIBLE_DEVICES'] = '-1'

print("=== LIGHTWEIGHT MNIST CNN CLASSIFICATION ===")
print("Optimized for i5 CPU with 15GB RAM")
print("="*50)

# =============================================
# STAGE a: Loading and preprocessing the image data
# =============================================

def load_and_preprocess_data_optimized():
    """
    Stage a: Optimized loading and preprocessing
    """
    print("\n" + "="*50)
    print("STAGE a: OPTIMIZED LOADING AND PREPROCESSING")
    print("="*50)
    
    # Load only first 20,000 samples for faster training (still plenty for MNIST)
    print("Loading training data from mnist_train.csv (first 20,000 samples)...")
    train_data = pd.read_csv('mnist_train.csv', nrows=20000)
    print("Loading test data from mnist_test.csv (first 5,000 samples)...")
    test_data = pd.read_csv('mnist_test.csv', nrows=5000)
    
    # Extract labels and images
    y_train = train_data.iloc[:, 0].values
    x_train = train_data.iloc[:, 1:].values
    
    y_test = test_data.iloc[:, 0].values
    x_test = test_data.iloc[:, 1:].values
    
    print(f"Raw training data shape: {x_train.shape}")
    print(f"Raw test data shape: {x_test.shape}")
    
    # Normalize pixel values
    x_train = x_train.astype('float32') / 255.0
    x_test = x_test.astype('float32') / 255.0
    
    # Reshape for CNN
    x_train = x_train.reshape(-1, 28, 28, 1)
    x_test = x_test.reshape(-1, 28, 28, 1)
    
    # Convert labels to categorical
    y_train_categorical = keras.utils.to_categorical(y_train, 10)
    y_test_categorical = keras.utils.to_categorical(y_test, 10)
    
    print(f"\nAfter preprocessing:")
    print(f"Training images: {x_train.shape}")
    print(f"Test images: {x_test.shape}")
    print(f"Label distribution: {np.bincount(y_train)}")
    
    return x_train, y_train_categorical, x_test, y_test_categorical, y_train, y_test

# =============================================
# STAGE b: Defining the model's architecture
# =============================================

def create_lightweight_cnn_model():
    """
    Stage b: Define lightweight CNN architecture
    """
    print("\n" + "="*50)
    print("STAGE b: LIGHTWEIGHT CNN ARCHITECTURE")
    print("="*50)
    
    model = keras.Sequential([
        # First Convolutional Block - LIGHTWEIGHT
        layers.Conv2D(16, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.2),
        
        # Second Convolutional Block - LIGHTWEIGHT  
        layers.Conv2D(32, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.2),
        
        # Flatten and Minimal Dense Layers
        layers.Flatten(),
        layers.Dense(64, activation='relu'),  # Reduced from 256
        layers.Dropout(0.3),
        
        # Output Layer
        layers.Dense(10, activation='softmax')
    ])
    
    # Compile with simpler optimizer
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    print("Lightweight CNN Model Architecture:")
    model.summary()
    print(f"Total parameters: {model.count_params():,} (Very lightweight!)")
    
    return model

# =============================================
# STAGE c: Training the model
# =============================================

def train_model_optimized(model, x_train, y_train, x_test, y_test):
    """
    Stage c: Optimized training
    """
    print("\n" + "="*50)
    print("STAGE c: OPTIMIZED TRAINING")
    print("="*50)
    
    # Simple callbacks only
    callbacks = [
        keras.callbacks.EarlyStopping(
            patience=3,  # Reduced patience
            restore_best_weights=True,
            monitor='val_accuracy'
        )
    ]
    
    print("Starting optimized training...")
    print(f"Training samples: {len(x_train)}")
    print(f"Validation samples: {len(x_test)}")
    print("Expected: 5-10 seconds per epoch on i5 CPU")
    
    # Train with larger batch size for speed
    history = model.fit(
        x_train, y_train,
        batch_size=256,  # Larger batch for speed
        epochs=10,       # Fewer epochs
        validation_data=(x_test, y_test),
        callbacks=callbacks,
        verbose=1,
        shuffle=True
    )
    
    print("Training completed successfully!")
    return history

# =============================================
# STAGE d: Estimating the model's performance
# =============================================

def evaluate_model_optimized(model, history, x_test, y_test, y_test_original):
    """
    Stage d: Optimized performance evaluation
    """
    print("\n" + "="*50)
    print("STAGE d: OPTIMIZED PERFORMANCE ESTIMATION")
    print("="*50)
    
    # Quick evaluation
    test_loss, test_accuracy = model.evaluate(x_test, y_test, verbose=0)
    
    print("📊 PERFORMANCE METRICS:")
    print("-" * 25)
    print(f"Test Accuracy: {test_accuracy:.4f}")
    print(f"Test Loss:     {test_loss:.4f}")
    
    # Quick training history
    plot_training_history_optimized(history)
    
    # Fast predictions
    predictions = model.predict(x_test, verbose=0, batch_size=512)
    predicted_classes = np.argmax(predictions, axis=1)
    true_classes = y_test_original
    
    overall_accuracy = np.mean(true_classes == predicted_classes)
    print(f"\n🎯 Overall Accuracy: {overall_accuracy:.4f}")
    
    # Quick confusion matrix
    plot_quick_confusion_matrix(true_classes, predicted_classes)
    
    # Quick classification report
    print("\n📈 QUICK CLASSIFICATION REPORT:")
    print("-" * 35)
    print(classification_report(true_classes, predicted_classes, digits=3))
    
    # Save model
    model.save('mnist_lightweight_cnn.h5')
    print(f"✓ Model saved as 'mnist_lightweight_cnn.h5'")
    
    return test_accuracy

def plot_training_history_optimized(history):
    """
    Quick training history plot
    """
    plt.figure(figsize=(10, 4))
    
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training', linewidth=2)
    plt.plot(history.history['val_accuracy'], label='Validation', linewidth=2)
    plt.title('Model Accuracy', fontweight='bold')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training', linewidth=2)
    plt.plot(history.history['val_loss'], label='Validation', linewidth=2)
    plt.title('Model Loss', fontweight='bold')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

def plot_quick_confusion_matrix(true_labels, pred_labels):
    """
    Quick confusion matrix
    """
    cm = confusion_matrix(true_labels, pred_labels)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title('Confusion Matrix', fontweight='bold')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.tight_layout()
    plt.show()

# =============================================
# MAIN EXECUTION - OPTIMIZED 4 STAGES
# =============================================

def main_optimized():
    """
    Execute optimized 4-stage pipeline
    """
    import time
    total_start = time.time()
    
    try:
        print("🚀 Starting optimized MNIST CNN classification...")
        print("💡 Optimized for: i5 CPU + 15GB RAM")
        print("⏱️  Expected total time: 1-2 minutes")
        
        # Stage a: Optimized loading
        stage_start = time.time()
        x_train, y_train, x_test, y_test, y_train_orig, y_test_orig = load_and_preprocess_data_optimized()
        print(f"✓ Stage a completed in {time.time() - stage_start:.1f}s")
        
        # Stage b: Lightweight model
        stage_start = time.time()
        model = create_lightweight_cnn_model()
        print(f"✓ Stage b completed in {time.time() - stage_start:.1f}s")
        
        # Stage c: Optimized training
        stage_start = time.time()
        history = train_model_optimized(model, x_train, y_train, x_test, y_test)
        training_time = time.time() - stage_start
        print(f"✓ Stage c completed in {training_time:.1f}s")
        print(f"⚡ Training speed: {training_time/len(history.history['accuracy']):.1f}s per epoch")
        
        # Stage d: Quick evaluation
        stage_start = time.time()
        final_accuracy = evaluate_model_optimized(model, history, x_test, y_test, y_test_orig)
        print(f"✓ Stage d completed in {time.time() - stage_start:.1f}s")
        
        total_time = time.time() - total_start
        
        # Final Summary
        print("\n" + "="*60)
        print("🎯 LIGHTWEIGHT CNN COMPLETED SUCCESSFULLY!")
        print("="*60)
        print(f"📊 Final Accuracy: {final_accuracy:.4f}")
        print(f"⏱️  Total Time: {total_time:.1f} seconds")
        print(f"📦 Model Size: {model.count_params():,} parameters")
        print(f"🔧 Optimized for: i5 CPU + 15GB RAM")
        print(f"✅ Perfect for practical exams!")
        print("="*60)
        
    except Exception as e:
        print(f"❌ Error: {e}")
        import traceback
        traceback.print_exc()

# Run the optimized pipeline
if __name__ == "__main__":
    main_optimized()
```
