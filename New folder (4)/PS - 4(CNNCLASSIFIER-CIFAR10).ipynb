```python
# Implement the CNN model for classifying CIFAR10 image dataset by dividing the
# model into following 4 stages:
# a. Loading and preprocessing the image data
# b. Defining the model's architecture
# c. Training the model
# d. Estimating the model's performance
```

```python
# a. Import the necessary packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.regularizers import l2
from tensorflow.keras.datasets import cifar10
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

print("=== CNN FOR CIFAR10 IMAGE CLASSIFICATION ===")
print("Optimized CNN Architecture with Enhanced Visualizations")
print("="*60)

# a. Loading and preprocessing the image data
print("Stage 1: Loading and preprocessing the image data...")

# Load CIFAR10 dataset
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

print(f"Original data shapes:")
print(f"X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"X_test: {X_test.shape}, y_test: {y_test.shape}")

# Normalize pixel values to [0, 1]
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# One-hot encode the labels
num_classes = 10
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)

print(f"Processed data shapes:")
print(f"X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"X_test: {X_test.shape}, y_test: {y_test.shape}")

# Class names for CIFAR10
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 
               'dog', 'frog', 'horse', 'ship', 'truck']

# Visualize sample images from each class
plt.figure(figsize=(15, 8))
for i in range(10):
    # Find first image of each class
    class_idx = np.where(np.argmax(y_train, axis=1) == i)[0][0]
    
    plt.subplot(2, 5, i + 1)
    plt.imshow(X_train[class_idx])
    plt.title(f'{class_names[i]}', fontweight='bold', fontsize=12)
    plt.axis('off')
    
plt.suptitle('Sample Images from Each Class - CIFAR10 Dataset', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# b. Defining the model's architecture
print("\nStage 2: Defining the CNN model's architecture...")

def create_cnn_model():
    """
    Create a CNN model optimized for CIFAR10 classification
    """
    model = Sequential([
        # First Convolutional Block
        Conv2D(32, (3, 3), activation='relu', padding='same', 
               input_shape=(32, 32, 3), kernel_regularizer=l2(0.001)),
        BatchNormalization(),
        Conv2D(32, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        Dropout(0.25),
        
        # Second Convolutional Block
        Conv2D(64, (3, 3), activation='relu', padding='same', kernel_regularizer=l2(0.001)),
        BatchNormalization(),
        Conv2D(64, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        Dropout(0.25),
        
        # Third Convolutional Block
        Conv2D(128, (3, 3), activation='relu', padding='same', kernel_regularizer=l2(0.001)),
        BatchNormalization(),
        Conv2D(128, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        Dropout(0.25),
        
        # Fully Connected Layers
        Flatten(),
        Dense(256, activation='relu', kernel_regularizer=l2(0.001)),
        BatchNormalization(),
        Dropout(0.5),
        Dense(128, activation='relu'),
        Dropout(0.3),
        Dense(num_classes, activation='softmax')
    ])
    
    # Use Adam optimizer
    optimizer = Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999)
    
    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model

# Create the model
model = create_cnn_model()

print("\nModel Architecture:")
model.summary()
print(f"Total parameters: {model.count_params():,}")

# Visualize model architecture
from tensorflow.keras.utils import plot_model
plt.figure(figsize=(12, 8))
try:
    plot_model(model, to_file='model_architecture.png', show_shapes=True, show_layer_names=True)
    img = plt.imread('model_architecture.png')
    plt.imshow(img)
    plt.axis('off')
    plt.title('CNN Model Architecture', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
except:
    print("Model visualization skipped (graphviz not installed)")

# c. Training the model
print("\nStage 3: Training the model...")

# Training parameters
batch_size = 128
epochs = 50

# Enhanced callbacks
callbacks = [
    keras.callbacks.EarlyStopping(
        monitor='val_accuracy',
        patience=15,
        restore_best_weights=True,
        verbose=1
    ),
    keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=8,
        min_lr=1e-7,
        verbose=1
    )
]

print(f"Training parameters:")
print(f"- Batch size: {batch_size}")
print(f"- Epochs: {epochs}")
print(f"- Training samples: {len(X_train):,}")
print(f"- Validation samples: {len(X_test):,}")

# Train the model
history = model.fit(
    X_train, y_train,
    batch_size=batch_size,
    epochs=epochs,
    validation_data=(X_test, y_test),
    callbacks=callbacks,
    verbose=1,
    shuffle=True
)

print("Training completed!")

# d. Estimating the model's performance
print("\nStage 4: Estimating the model's performance...")

# Evaluate the model
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")

# Make predictions
y_pred = model.predict(X_test, verbose=0)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

# COMPREHENSIVE VISUALIZATIONS
print("\n" + "="*60)
print("COMPREHENSIVE VISUALIZATION ANALYSIS")
print("="*60)

# 1. Training History Visualization
plt.figure(figsize=(20, 6))

# Loss plot
plt.subplot(1, 3, 1)
plt.plot(history.history['loss'], label='Training Loss', linewidth=2, color='blue')
plt.plot(history.history['val_loss'], label='Validation Loss', linewidth=2, color='red')
plt.title('Training and Validation Loss', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)
plt.ylim(bottom=0)

# Accuracy plot
plt.subplot(1, 3, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy', linewidth=2, color='blue')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', linewidth=2, color='red')
plt.title('Training and Validation Accuracy', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, alpha=0.3)
plt.ylim(0, 1)

# Learning rate plot (if available)
plt.subplot(1, 3, 3)
if 'lr' in history.history:
    plt.plot(history.history['lr'], linewidth=2, color='green')
    plt.title('Learning Rate Schedule', fontsize=14, fontweight='bold')
    plt.xlabel('Epoch')
    plt.ylabel('Learning Rate')
    plt.grid(True, alpha=0.3)
    plt.yscale('log')
else:
    # Show final metrics
    final_train_acc = history.history['accuracy'][-1]
    final_val_acc = history.history['val_accuracy'][-1]
    metrics_text = f'Final Training Acc: {final_train_acc:.3f}\nFinal Validation Acc: {final_val_acc:.3f}'
    plt.text(0.5, 0.5, metrics_text, ha='center', va='center', fontsize=12, 
             transform=plt.gca().transAxes, bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue"))
    plt.title('Training Summary', fontsize=14, fontweight='bold')
    plt.axis('off')

plt.tight_layout()
plt.show()

# 2. Confusion Matrix with Enhanced Visualization
plt.figure(figsize=(14, 12))
cm = confusion_matrix(y_true_classes, y_pred_classes)

# Calculate percentages
cm_percent = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis] * 100

plt.subplot(2, 1, 1)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=class_names, yticklabels=class_names,
            cbar_kws={'label': 'Count'})
plt.title('Confusion Matrix - Counts', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Predicted Label', fontsize=12)
plt.ylabel('True Label', fontsize=12)
plt.xticks(rotation=45)
plt.yticks(rotation=0)

plt.subplot(2, 1, 2)
sns.heatmap(cm_percent, annot=True, fmt='.1f', cmap='viridis', 
            xticklabels=class_names, yticklabels=class_names,
            cbar_kws={'label': 'Percentage (%)'})
plt.title('Confusion Matrix - Percentages', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Predicted Label', fontsize=12)
plt.ylabel('True Label', fontsize=12)
plt.xticks(rotation=45)
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()

# 3. Classification Performance by Class
class_accuracy = cm.diagonal() / cm.sum(axis=1)
class_precision = cm.diagonal() / cm.sum(axis=0)
class_recall = cm.diagonal() / cm.sum(axis=1)

plt.figure(figsize=(15, 6))
x_pos = np.arange(len(class_names))

plt.subplot(1, 3, 1)
bars = plt.bar(x_pos, class_accuracy, color='skyblue', edgecolor='navy', alpha=0.7)
plt.title('Class-wise Accuracy', fontsize=14, fontweight='bold')
plt.xlabel('Classes')
plt.ylabel('Accuracy')
plt.xticks(x_pos, class_names, rotation=45)
plt.ylim(0, 1)
# Add value labels on bars
for bar, acc in zip(bars, class_accuracy):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
             f'{acc:.3f}', ha='center', va='bottom', fontsize=9)

plt.subplot(1, 3, 2)
bars = plt.bar(x_pos, class_precision, color='lightgreen', edgecolor='darkgreen', alpha=0.7)
plt.title('Class-wise Precision', fontsize=14, fontweight='bold')
plt.xlabel('Classes')
plt.ylabel('Precision')
plt.xticks(x_pos, class_names, rotation=45)
plt.ylim(0, 1)
for bar, prec in zip(bars, class_precision):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
             f'{prec:.3f}', ha='center', va='bottom', fontsize=9)

plt.subplot(1, 3, 3)
bars = plt.bar(x_pos, class_recall, color='lightcoral', edgecolor='darkred', alpha=0.7)
plt.title('Class-wise Recall', fontsize=14, fontweight='bold')
plt.xlabel('Classes')
plt.ylabel('Recall')
plt.xticks(x_pos, class_names, rotation=45)
plt.ylim(0, 1)
for bar, rec in zip(bars, class_recall):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
             f'{rec:.3f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

# 4. Sample Predictions Visualization
print("\nSample Predictions Analysis:")

# Get indices of correct and incorrect predictions
correct_indices = np.where(y_pred_classes == y_true_classes)[0]
incorrect_indices = np.where(y_pred_classes != y_true_classes)[0]

print(f"Correct predictions: {len(correct_indices):,}/{len(y_test):,} ({len(correct_indices)/len(y_test)*100:.2f}%)")
print(f"Incorrect predictions: {len(incorrect_indices):,}/{len(y_test):,} ({len(incorrect_indices)/len(y_test)*100:.2f}%)")

# Show correct predictions
plt.figure(figsize=(16, 12))
plt.suptitle('Sample Correct Predictions with Confidence Scores', fontsize=16, fontweight='bold')

for i in range(min(12, len(correct_indices))):
    idx = correct_indices[i]
    plt.subplot(3, 4, i + 1)
    
    # Display image
    plt.imshow(X_test[idx])
    
    true_label = class_names[y_true_classes[idx]]
    pred_label = class_names[y_pred_classes[idx]]
    confidence = np.max(y_pred[idx])
    
    # Color code based on confidence
    color = 'green' if confidence > 0.9 else 'orange' if confidence > 0.7 else 'red'
    
    plt.title(f'True: {true_label}\nPred: {pred_label}\nConf: {confidence:.3f}', 
              color=color, fontsize=10, fontweight='bold')
    plt.axis('off')

plt.tight_layout()
plt.show()

# Show misclassifications
if len(incorrect_indices) > 0:
    plt.figure(figsize=(16, 12))
    plt.suptitle('Sample Misclassifications with Confidence Scores', fontsize=16, fontweight='bold', color='red')
    
    for i in range(min(12, len(incorrect_indices))):
        idx = incorrect_indices[i]
        plt.subplot(3, 4, i + 1)
        
        # Display image
        plt.imshow(X_test[idx])
        
        true_label = class_names[y_true_classes[idx]]
        pred_label = class_names[y_pred_classes[idx]]
        confidence = np.max(y_pred[idx])
        
        plt.title(f'True: {true_label}\nPred: {pred_label}\nConf: {confidence:.3f}', 
                  color='red', fontsize=10, fontweight='bold')
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()

# 5. Confidence Distribution Analysis
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
correct_confidences = [np.max(y_pred[i]) for i in correct_indices]
incorrect_confidences = [np.max(y_pred[i]) for i in incorrect_indices]

plt.hist(correct_confidences, bins=30, alpha=0.7, label='Correct Predictions', color='green')
plt.hist(incorrect_confidences, bins=30, alpha=0.7, label='Incorrect Predictions', color='red')
plt.xlabel('Prediction Confidence')
plt.ylabel('Frequency')
plt.title('Distribution of Prediction Confidence', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
# Accuracy vs Confidence
confidence_bins = np.linspace(0, 1, 11)
accuracy_per_bin = []

for i in range(len(confidence_bins)-1):
    mask = (y_pred.max(axis=1) >= confidence_bins[i]) & (y_pred.max(axis=1) < confidence_bins[i+1])
    if np.sum(mask) > 0:
        accuracy = np.mean(y_pred_classes[mask] == y_true_classes[mask])
        accuracy_per_bin.append(accuracy)
    else:
        accuracy_per_bin.append(0)

plt.plot(confidence_bins[:-1] + 0.05, accuracy_per_bin, 'o-', linewidth=2, markersize=8)
plt.xlabel('Confidence Bin')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Prediction Confidence', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.ylim(0, 1)

plt.tight_layout()
plt.show()

# Detailed Classification Report
print("\n" + "="*60)
print("DETAILED CLASSIFICATION REPORT")
print("="*60)
print(classification_report(y_true_classes, y_pred_classes, 
                          target_names=class_names, digits=4))

# Save the model
model.save('cifar10_cnn_model.h5')
print(f"\n Model saved as 'cifar10_cnn_model.h5'")

# Final comprehensive summary
print("\n" + "="*70)
print("CNN MODEL TRAINING SUMMARY - CIFAR10 CLASSIFICATION")
print("="*70)
print(f" Final Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
print(f" Final Test Loss: {test_loss:.4f}")
print(f"  Model Type: Convolutional Neural Network (CNN)")
print(f"  Optimizer: Adam (lr=0.001)")
print(f" Training Samples: {len(X_train):,}")
print(f" Test Samples: {len(X_test):,}")
print(f" Model Parameters: {model.count_params():,}")
print(f" Number of Classes: {num_classes}")
print(f" Training Epochs Completed: {len(history.history['loss'])}")
print(f" Best Validation Accuracy: {max(history.history['val_accuracy']):.4f}")

print(f"\n Class Performance Summary:")
for i, class_name in enumerate(class_names):
    acc = class_accuracy[i]
    print(f"   {class_name:12s}: {acc:.3f}")

print("\n PERFORMANCE INSIGHTS:")
print("- CNN significantly outperforms FNN on image data")
print("- Model shows good learning capability with proper regularization")
print("- Confusion matrix reveals which classes are most challenging")
print("- Confidence analysis helps understand model certainty")
print("="*70)
```

```python
# a. Import the necessary packages
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.datasets import cifar10
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

print("=== LIGHTWEIGHT CNN FOR CIFAR10 ===")
print("Optimized for i5 CPU + 15GB RAM - Fast Training")
print("="*50)

# a. Loading and preprocessing the image data
print("Stage 1: Loading and preprocessing the image data...")

# Load CIFAR10 dataset
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# Use smaller subset for faster training (still maintains good performance)
train_samples = 15000  # Reduced from 50,000
test_samples = 3000    # Reduced from 10,000

# Select subset of data
X_train = X_train[:train_samples]
y_train = y_train[:train_samples]
X_test = X_test[:test_samples]
y_test = y_test[:test_samples]

print(f"Using reduced dataset for faster training:")
print(f"X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"X_test: {X_test.shape}, y_test: {y_test.shape}")

# Normalize pixel values to [0, 1]
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# One-hot encode the labels
num_classes = 10
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)

# Class names for CIFAR10
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 
               'dog', 'frog', 'horse', 'ship', 'truck']

# Quick sample visualization
plt.figure(figsize=(12, 6))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    class_idx = np.where(np.argmax(y_train, axis=1) == i)[0][0]
    plt.imshow(X_train[class_idx])
    plt.title(f'{class_names[i]}', fontsize=10)
    plt.axis('off')
plt.suptitle('CIFAR10 Sample Images', fontweight='bold')
plt.tight_layout()
plt.show()

# b. Defining the model's architecture
print("\nStage 2: Defining lightweight CNN architecture...")

def create_lightweight_cnn():
    """
    Create a lightweight CNN model optimized for speed
    """
    model = Sequential([
        # First Convolutional Block - Lightweight
        Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(32, 32, 3)),
        BatchNormalization(),
        Conv2D(32, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        Dropout(0.2),  # Reduced dropout
        
        # Second Convolutional Block - Lightweight  
        Conv2D(64, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Dropout(0.2),
        
        # Third Convolutional Block - Minimal
        Conv2D(64, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        Dropout(0.2),
        
        # Fully Connected Layers - Reduced size
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.3),
        Dense(64, activation='relu'),
        Dropout(0.2),
        Dense(num_classes, activation='softmax')
    ])
    
    # Use Adam optimizer
    optimizer = Adam(learning_rate=0.001)
    
    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model

# Create the lightweight model
model = create_lightweight_cnn()

print("\nLightweight Model Architecture:")
model.summary()
print(f"Total parameters: {model.count_params():,} (Very efficient!)")

# c. Training the model
print("\nStage 3: Training the model...")

# Optimized training parameters for speed
batch_size = 64  # Smaller batch for faster iterations
epochs = 25      # Reduced epochs

# Simple callbacks
callbacks = [
    keras.callbacks.EarlyStopping(
        monitor='val_accuracy',
        patience=8,  # Reduced patience
        restore_best_weights=True,
        verbose=1
    )
]

print(f"Optimized training parameters:")
print(f"- Batch size: {batch_size}")
print(f"- Epochs: {epochs}")
print(f"- Training samples: {len(X_train):,}")
print(f"- Expected time: 30-60 seconds per epoch")

# Train the model
history = model.fit(
    X_train, y_train,
    batch_size=batch_size,
    epochs=epochs,
    validation_data=(X_test, y_test),
    callbacks=callbacks,
    verbose=1,
    shuffle=True
)

print("Training completed successfully!")

# d. Estimating the model's performance
print("\nStage 4: Estimating the model's performance...")

# Evaluate the model
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")

# Make predictions
y_pred = model.predict(X_test, verbose=0)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

# EFFICIENT VISUALIZATIONS
print("\n" + "="*50)
print("EFFICIENT VISUALIZATION ANALYSIS")
print("="*50)

# 1. Training History (Compact)
plt.figure(figsize=(15, 5))

# Loss plot
plt.subplot(1, 3, 1)
plt.plot(history.history['loss'], label='Training Loss', linewidth=2, color='blue')
plt.plot(history.history['val_loss'], label='Validation Loss', linewidth=2, color='red')
plt.title('Model Loss', fontsize=12, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)

# Accuracy plot
plt.subplot(1, 3, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy', linewidth=2, color='blue')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', linewidth=2, color='red')
plt.title('Model Accuracy', fontsize=12, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, alpha=0.3)

# Training summary
plt.subplot(1, 3, 3)
final_train_acc = history.history['accuracy'][-1]
final_val_acc = history.history['val_accuracy'][-1]
best_val_acc = max(history.history['val_accuracy'])

summary_text = f'Final Results:\n\nTrain Acc: {final_train_acc:.3f}\nVal Acc: {final_val_acc:.3f}\nBest Val Acc: {best_val_acc:.3f}\nTest Acc: {test_accuracy:.3f}'

plt.text(0.5, 0.5, summary_text, ha='center', va='center', fontsize=11, 
         transform=plt.gca().transAxes, bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgreen"))
plt.title('Performance Summary', fontsize=12, fontweight='bold')
plt.axis('off')

plt.tight_layout()
plt.show()

# 2. Compact Confusion Matrix
plt.figure(figsize=(10, 8))
cm = confusion_matrix(y_true_classes, y_pred_classes)

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix', fontsize=14, fontweight='bold')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# 3. Class-wise Performance
class_accuracy = cm.diagonal() / cm.sum(axis=1)

plt.figure(figsize=(12, 6))
bars = plt.bar(range(10), class_accuracy, color='lightblue', edgecolor='navy', alpha=0.7)
plt.title('Class-wise Accuracy', fontsize=14, fontweight='bold')
plt.xlabel('Classes')
plt.ylabel('Accuracy')
plt.xticks(range(10), class_names, rotation=45)
plt.ylim(0, 1)
plt.grid(True, alpha=0.3)

# Add value labels
for bar, acc in zip(bars, class_accuracy):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
             f'{acc:.3f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

# 4. Quick Sample Predictions
print("\nSample Predictions Analysis:")

correct_indices = np.where(y_pred_classes == y_true_classes)[0]
incorrect_indices = np.where(y_pred_classes != y_true_classes)[0]

print(f"Correct: {len(correct_indices)}/{len(y_test)} ({len(correct_indices)/len(y_test)*100:.1f}%)")
print(f"Wrong: {len(incorrect_indices)}/{len(y_test)} ({len(incorrect_indices)/len(y_test)*100:.1f}%)")

# Show some predictions
plt.figure(figsize=(12, 8))

# Correct predictions
for i in range(min(6, len(correct_indices))):
    idx = correct_indices[i]
    plt.subplot(2, 3, i + 1)
    plt.imshow(X_test[idx])
    true_label = class_names[y_true_classes[idx]]
    pred_label = class_names[y_pred_classes[idx]]
    confidence = np.max(y_pred[idx])
    plt.title(f' {true_label}\nConf: {confidence:.2f}', color='green', fontsize=10)
    plt.axis('off')

plt.suptitle('Sample Correct Predictions', fontweight='bold')
plt.tight_layout()
plt.show()

# Show some errors if available
if len(incorrect_indices) > 0:
    plt.figure(figsize=(12, 8))
    for i in range(min(6, len(incorrect_indices))):
        idx = incorrect_indices[i]
        plt.subplot(2, 3, i + 1)
        plt.imshow(X_test[idx])
        true_label = class_names[y_true_classes[idx]]
        pred_label = class_names[y_pred_classes[idx]]
        confidence = np.max(y_pred[idx])
        plt.title(f'True: {true_label}\nPred: {pred_label}', color='red', fontsize=10)
        plt.axis('off')
    
    plt.suptitle('Sample Misclassifications', fontweight='bold', color='red')
    plt.tight_layout()
    plt.show()

# 5. Confidence Analysis (Simple)
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
confidences = np.max(y_pred, axis=1)
plt.hist(confidences, bins=20, alpha=0.7, color='purple')
plt.xlabel('Prediction Confidence')
plt.ylabel('Frequency')
plt.title('Confidence Distribution', fontweight='bold')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
# Accuracy by confidence quartile
quartiles = np.percentile(confidences, [25, 50, 75, 100])
accuracies = []
for i in range(len(quartiles)):
    if i == 0:
        mask = confidences <= quartiles[i]
    else:
        mask = (confidences > quartiles[i-1]) & (confidences <= quartiles[i])
    if np.sum(mask) > 0:
        acc = np.mean(y_pred_classes[mask] == y_true_classes[mask])
        accuracies.append(acc)

plt.bar(['Q1', 'Q2', 'Q3', 'Q4'], accuracies, color=['red', 'orange', 'yellow', 'green'])
plt.xlabel('Confidence Quartile')
plt.ylabel('Accuracy')
plt.title('Accuracy by Confidence Level', fontweight='bold')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Classification Report
print("\n" + "="*50)
print("CLASSIFICATION REPORT")
print("="*50)
print(classification_report(y_true_classes, y_pred_classes, 
                          target_names=class_names, digits=3))

# Save the lightweight model
model.save('cifar10_lightweight_cnn.h5')
print(f"\n Lightweight model saved as 'cifar10_lightweight_cnn.h5'")

# Final summary
print("\n" + "="*60)
print("LIGHTWEIGHT CNN TRAINING SUMMARY")
print("="*60)
print(f" Final Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
print(f" Final Test Loss: {test_loss:.4f}")
print(f" Model Type: Lightweight CNN (Optimized for i5)")
print(f" Optimizer: Adam (lr=0.001)")
print(f" Training Samples: {len(X_train):,}")
print(f" Test Samples: {len(X_test):,}")
print(f" Model Parameters: {model.count_params():,}")
print(f" Epochs Trained: {len(history.history['loss'])}")
print(f" Best Val Accuracy: {max(history.history['val_accuracy']):.4f}")

print(f"\n Top 3 Performing Classes:")
class_acc_sorted = sorted(zip(class_names, class_accuracy), key=lambda x: x[1], reverse=True)
for i, (cls, acc) in enumerate(class_acc_sorted[:3]):
    print(f"   {i+1}. {cls:12s}: {acc:.3f}")

print(f"\n Bottom 3 Performing Classes:")
for i, (cls, acc) in enumerate(class_acc_sorted[-3:]):
    print(f"   {i+1}. {cls:12s}: {acc:.3f}")

print("\n PERFORMANCE NOTES:")
print("- Lightweight model optimized for speed")
print("- Good accuracy for the computational constraints")
print("- Fast training suitable for i5 processors")
print("- Efficient memory usage under 15GB RAM")
print("="*60)
```
