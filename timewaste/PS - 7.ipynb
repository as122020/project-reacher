```python
# ============================================================
# Continuous Bag of Words (CBOW) Model Implementation
# With Real Text Document Loading
# ============================================================

# a. Import required libraries
import tensorflow as tf
from tensorflow.keras import models, layers, optimizers
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import re
import os
from sklearn.manifold import TSNE
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

# Download NLTK data
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')

try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('stopwords')

print(f"TensorFlow version: {tf.__version__}")

# For reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# ============================================================
# a. Data Preparation with Real Text Documents
# ============================================================

print("=" * 60)
print("STAGE A: DATA PREPARATION WITH REAL TEXT DOCUMENTS")
print("=" * 60)

def create_sample_text_document():
    """
    Create a comprehensive sample text document about AI and Machine Learning
    """
    sample_text = """
    Artificial Intelligence and Machine Learning Fundamentals
    
    Artificial intelligence is the simulation of human intelligence processes by machines. 
    These processes include learning, reasoning, and self-correction. Machine learning 
    is a subset of artificial intelligence that provides systems the ability to automatically 
    learn and improve from experience without being explicitly programmed.
    
    Deep learning is a specialized form of machine learning that uses neural networks 
    with multiple layers. These deep neural networks can learn and make intelligent 
    decisions on their own. Natural language processing enables computers to understand, 
    interpret, and generate human language in a valuable way.
    
    Computer vision is another important field of artificial intelligence that enables 
    computers to identify and process objects in images and videos. Reinforcement learning 
    is a type of machine learning where an agent learns to make decisions by taking 
    actions in an environment to achieve maximum cumulative reward.
    
    Supervised learning involves training a model on labeled data, while unsupervised 
    learning finds hidden patterns in unlabeled data. Semi-supervised learning combines 
    both labeled and unlabeled data during training. The training process involves 
    feeding data to algorithms and allowing them to learn patterns and relationships.
    
    Neural networks consist of interconnected nodes organized in layers. Each connection 
    can transmit signals to other neurons. The network learns by adjusting the weights 
    of these connections. Backpropagation is the algorithm used for training neural 
    networks by calculating gradients of the loss function.
    
    Convolutional neural networks are particularly effective for image recognition 
    tasks, while recurrent neural networks excel at sequence prediction problems. 
    Transformers have revolutionized natural language processing with their attention 
    mechanisms that weigh the importance of different words in a sequence.
    
    Data preprocessing is crucial for machine learning projects. This includes 
    cleaning data, handling missing values, and feature engineering. Feature 
    selection helps in choosing the most relevant attributes for model training.
    
    Model evaluation metrics include accuracy, precision, recall, and F1-score 
    for classification tasks. For regression problems, common metrics are mean 
    squared error and R-squared. Cross-validation ensures that model performance 
    is consistent across different data splits.
    
    Overfitting occurs when a model learns the training data too well but fails 
    to generalize to new data. Regularization techniques like L1 and L2 help 
    prevent overfitting by adding penalty terms to the loss function.
    
    The future of artificial intelligence holds tremendous potential across 
    various industries including healthcare, finance, transportation, and 
    education. Ethical considerations and responsible AI development are 
    becoming increasingly important as these technologies advance.
    """
    return sample_text

def load_technical_document():
    """
    Load a technical document about neural networks and deep learning
    """
    technical_text = """
    Neural Networks and Deep Learning Architecture
    
    A neural network is composed of layers of interconnected neurons. The input layer 
    receives the initial data, hidden layers process the information, and the output 
    layer produces the final result. Each neuron applies an activation function to 
    its input, introducing non-linearity into the network.
    
    Activation functions determine whether a neuron should be activated. Common 
    activation functions include sigmoid, tanh, ReLU, and softmax. The ReLU 
    function has become popular due to its computational efficiency and ability 
    to mitigate the vanishing gradient problem.
    
    The loss function measures how well the neural network performs on the training 
    data. For classification tasks, cross-entropy loss is commonly used, while 
    mean squared error is typical for regression problems. The optimizer adjusts 
    the weights to minimize the loss function during training.
    
    Gradient descent is the fundamental optimization algorithm used in training 
    neural networks. Stochastic gradient descent updates weights for each training 
    example, while batch gradient descent uses the entire dataset. Mini-batch 
    gradient descent strikes a balance by using small batches of data.
    
    Convolutional Neural Networks are designed for processing grid-like data 
    such as images. They use convolutional layers that apply filters to detect 
    features, pooling layers that reduce spatial dimensions, and fully connected 
    layers for final classification.
    
    Recurrent Neural Networks handle sequential data by maintaining internal 
    state. Long Short-Term Memory networks and Gated Recurrent Units are RNN 
    variants that address the vanishing gradient problem and can learn long-term 
    dependencies in sequence data.
    
    Attention mechanisms allow neural networks to focus on relevant parts of 
    the input when making predictions. Transformer architectures use self-attention 
    to process all parts of the input simultaneously rather than sequentially.
    
    Transfer learning enables using pre-trained models on new tasks with limited 
    data. Fine-tuning adjusts the weights of a pre-trained model for specific 
    applications. This approach has proven particularly effective in computer 
    vision and natural language processing.
    
    Hyperparameter tuning involves finding the optimal settings for learning 
    rate, batch size, number of layers, and other model parameters. Techniques 
    include grid search, random search, and Bayesian optimization.
    
    Model deployment involves serving trained models in production environments. 
    Monitoring model performance over time is crucial as data distributions may 
    change, requiring model retraining or updating.
    """
    return technical_text

def load_research_paper_abstract():
    """
    Load research paper abstracts about machine learning
    """
    research_text = """
    Machine Learning Research and Applications
    
    Recent advances in machine learning have transformed numerous industries and 
    research domains. Deep learning models have achieved state-of-the-art performance 
    in image recognition, speech processing, and natural language understanding. 
    These models learn hierarchical representations from raw data through multiple 
    processing layers.
    
    Generative adversarial networks consist of two neural networks, a generator 
    and a discriminator, that compete against each other. The generator creates 
    synthetic data while the discriminator evaluates its authenticity. This 
    adversarial training process produces highly realistic synthetic data.
    
    Autoencoders are neural networks designed for unsupervised learning of efficient 
    data representations. They consist of an encoder that compresses the input 
    and a decoder that reconstructs the input from the compressed representation. 
    Variational autoencoders add probabilistic elements to enable generative modeling.
    
    Graph neural networks operate on graph-structured data, making them suitable 
    for social network analysis, molecular chemistry, and recommendation systems. 
    They aggregate information from neighboring nodes to learn node representations.
    
    Meta-learning, or learning to learn, focuses on developing algorithms that 
    can adapt to new tasks with limited data. Few-shot learning aims to recognize 
    new classes from only a few examples by leveraging knowledge from related tasks.
    
    Explainable AI methods provide insights into model predictions, increasing 
    transparency and trust in machine learning systems. Techniques include 
    feature importance scores, attention visualization, and counterfactual explanations.
    
    Federated learning enables model training across decentralized devices while 
    keeping data localized. This approach preserves privacy and reduces data 
    transmission costs while still benefiting from collective learning.
    
    Self-supervised learning leverages unlabeled data by creating supervisory 
    signals from the data itself. This approach has shown remarkable success 
    in natural language processing and computer vision tasks.
    
    Multimodal learning combines information from different data modalities such 
    as text, images, and audio. Cross-modal transformers can process and relate 
    information across different types of data simultaneously.
    
    The integration of machine learning with other technologies like blockchain, 
    Internet of Things, and edge computing is creating new opportunities and 
    challenges for researchers and practitioners in the field.
    """
    return research_text

def get_text_document_choice():
    """
    Let user choose which text document to use
    """
    print("\nğŸ“š AVAILABLE TEXT DOCUMENTS:")
    print("1. AI and Machine Learning Fundamentals (Comprehensive)")
    print("2. Neural Networks Technical Documentation")
    print("3. Research Paper Abstracts Collection")
    print("4. Load from file")
    
    choice = input("\nChoose text document (1-4): ").strip()
    
    if choice == '1':
        return create_sample_text_document(), "AI and Machine Learning Fundamentals"
    elif choice == '2':
        return load_technical_document(), "Neural Networks Technical Documentation"
    elif choice == '3':
        return load_research_paper_abstract(), "Research Paper Abstracts Collection"
    elif choice == '4':
        file_path = input("Enter file path: ").strip()
        try:
            with open(file_path, 'r', encoding='utf-8') as file:
                return file.read(), f"Custom File: {file_path}"
        except FileNotFoundError:
            print("âŒ File not found. Using default document.")
            return create_sample_text_document(), "AI and Machine Learning Fundamentals"
    else:
        print("âŒ Invalid choice. Using default document.")
        return create_sample_text_document(), "AI and Machine Learning Fundamentals"

def preprocess_text(text, remove_stopwords=True, min_word_length=2):
    """
    Preprocess the text with configurable options
    """
    # Convert to lowercase
    text = text.lower()
    
    # Remove punctuation and numbers
    text = re.sub(r'[^\w\s]', ' ', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'\s+', ' ', text)  # Remove extra whitespace
    
    # Tokenize
    tokens = word_tokenize(text)
    
    if remove_stopwords:
        # Remove stopwords
        stop_words = set(stopwords.words('english'))
        tokens = [token for token in tokens if token not in stop_words and len(token) >= min_word_length]
    else:
        tokens = [token for token in tokens if len(token) >= min_word_length]
    
    return tokens

# Load the chosen text document
print("\nğŸ”„ LOADING TEXT DOCUMENT...")
text_data, document_name = get_text_document_choice()

print(f"\nğŸ“„ DOCUMENT: {document_name}")
print(f"ğŸ“Š Document length: {len(text_data):,} characters")
print(f"ğŸ“ Preview (first 300 characters):")
print(text_data[:300] + "..." if len(text_data) > 300 else text_data)

# Preprocessing configuration
print("\nâš™ï¸  PREPROCESSING CONFIGURATION:")
remove_stopwords = input("Remove stopwords? (y/n, default y): ").strip().lower() != 'n'
min_length = input("Minimum word length (default 2): ").strip()
min_word_length = int(min_length) if min_length.isdigit() else 2

# Preprocess the text
tokens = preprocess_text(text_data, remove_stopwords, min_word_length)

print(f"\nğŸ“Š TEXT PROCESSING STATISTICS:")
print(f"Total tokens: {len(tokens):,}")
print(f"Unique tokens: {len(set(tokens)):,}")
print(f"Tokenized sample (first 25): {tokens[:25]}")

# Build vocabulary
word_freq = Counter(tokens)
vocab = sorted(word_freq, key=word_freq.get, reverse=True)
word_to_index = {word: idx for idx, word in enumerate(vocab)}
index_to_word = {idx: word for idx, word in enumerate(vocab)}

vocab_size = len(vocab)
print(f"\nğŸ“š VOCABULARY ANALYSIS:")
print(f"Vocabulary size: {vocab_size}")
print(f"Most common words (top 15):")
for word, freq in word_freq.most_common(15):
    print(f"  {word}: {freq}")

# Vocabulary coverage analysis
total_tokens = len(tokens)
covered_tokens = sum(word_freq[word] for word in vocab[:50])  # Top 50 words
coverage_percentage = (covered_tokens / total_tokens) * 100
print(f"Top 50 words cover {coverage_percentage:.1f}% of all tokens")

# ============================================================
# b. Generate Training Data
# ============================================================

print("\n" + "=" * 60)
print("STAGE B: GENERATE TRAINING DATA")
print("=" * 60)

def generate_cbow_data(tokens, word_to_index, window_size=2):
    """
    Generate training data for CBOW model
    """
    X = []  # Context words
    y = []  # Target words
    
    for i in range(window_size, len(tokens) - window_size):
        # Get context words (surrounding words)
        context = []
        for j in range(-window_size, window_size + 1):
            if j != 0:  # Skip the target word itself
                context.append(tokens[i + j])
        
        # Get target word
        target = tokens[i]
        
        # Convert to indices
        context_indices = [word_to_index[word] for word in context if word in word_to_index]
        
        # Only add if we have all context words
        if len(context_indices) == window_size * 2:
            target_index = word_to_index[target]
            X.append(context_indices)
            y.append(target_index)
    
    return np.array(X), np.array(y)

# Get window size from user
window_input = input("Enter context window size (default 2): ").strip()
window_size = int(window_input) if window_input.isdigit() else 2

print(f"\nğŸ”„ GENERATING TRAINING DATA WITH WINDOW SIZE {window_size}...")

# Generate training data
X, y = generate_cbow_data(tokens, word_to_index, window_size)

print(f"ğŸ“Š TRAINING DATA GENERATED:")
print(f"Context words (X) shape: {X.shape}")
print(f"Target words (y) shape: {y.shape}")
print(f"Window size: {window_size}")
print(f"Context words per sample: {X.shape[1]}")
print(f"Total training samples: {len(X):,}")

# Display sample training data with context
print(f"\nğŸ” SAMPLE TRAINING EXAMPLES:")
print(f"Format: [left context] <TARGET> [right context]")
for i in range(min(8, len(X))):
    context_indices = X[i]
    target_index = y[i]
    
    # Reconstruct the context window
    left_context = [index_to_word[idx] for idx in context_indices[:window_size]]
    right_context = [index_to_word[idx] for idx in context_indices[window_size:]]
    target_word = index_to_word[target_index]
    
    print(f"Example {i+1}: {left_context} <{target_word}> {right_context}")

# ============================================================
# c. Train Model
# ============================================================

print("\n" + "=" * 60)
print("STAGE C: TRAIN CBOW MODEL")
print("=" * 60)

def build_cbow_model(vocab_size, embedding_dim=100, window_size=2):
    """
    Build CBOW model architecture
    """
    model = models.Sequential([
        # Embedding layer for context words
        layers.Embedding(input_dim=vocab_size, 
                        output_dim=embedding_dim, 
                        input_length=window_size * 2,
                        name='embedding'),
        
        # Average the embeddings of context words
        layers.GlobalAveragePooling1D(name='average_pooling'),
        
        # Dense layers with ReLU activation
        layers.Dense(256, activation='relu', name='hidden_layer_1'),
        layers.Dropout(0.3),
        layers.Dense(128, activation='relu', name='hidden_layer_2'),
        layers.Dropout(0.3),
        
        # Output layer with softmax for word prediction
        layers.Dense(vocab_size, activation='softmax', name='output')
    ])
    
    return model

# Get model parameters
print("\nâš™ï¸  MODEL CONFIGURATION:")
embedding_input = input("Enter embedding dimension (default 100): ").strip()
embedding_dim = int(embedding_input) if embedding_input.isdigit() else 100

epochs_input = input("Enter number of epochs (default 50): ").strip()
epochs = int(epochs_input) if epochs_input.isdigit() else 50

batch_input = input("Enter batch size (default 32): ").strip()
batch_size = int(batch_input) if batch_input.isdigit() else 32

# Build model
print(f"\nğŸ”§ BUILDING CBOW MODEL...")
cbow_model = build_cbow_model(vocab_size, embedding_dim, window_size)

print("ğŸ“ MODEL ARCHITECTURE:")
cbow_model.summary()

# Compile the model
cbow_model.compile(
    optimizer=optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("\nâœ… MODEL COMPILED SUCCESSFULLY:")
print(f"Optimizer: Adam (learning_rate=0.001)")
print(f"Loss: Sparse Categorical Crossentropy")
print(f"Metrics: Accuracy")
print(f"Embedding dimension: {embedding_dim}")
print(f"Training epochs: {epochs}")
print(f"Batch size: {batch_size}")

# Train the model
print(f"\nğŸš€ STARTING MODEL TRAINING...")
history = cbow_model.fit(
    X, y,
    epochs=epochs,
    batch_size=batch_size,
    validation_split=0.2,
    verbose=1,
    shuffle=True
)

print("\nğŸ¯ TRAINING COMPLETED!")

# ============================================================
# d. Output and Results
# ============================================================

print("\n" + "=" * 60)
print("STAGE D: OUTPUT AND EVALUATION")
print("=" * 60)

# 1. Training Performance Visualization
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss', linewidth=2, color='blue')
plt.plot(history.history['val_loss'], label='Validation Loss', linewidth=2, color='red')
plt.title('CBOW Model - Training and Validation Loss', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy', linewidth=2, color='green')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', linewidth=2, color='orange')
plt.title('CBOW Model - Training and Validation Accuracy', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 2. Extract and analyze word embeddings
embedding_layer = cbow_model.get_layer('embedding')
weights = embedding_layer.get_weights()[0]
print(f"\nğŸ“Š WORD EMBEDDINGS EXTRACTED:")
print(f"Embedding matrix shape: {weights.shape}")
print(f"Total word vectors: {weights.shape[0]}")
print(f"Vector dimension: {weights.shape[1]}")

# 3. Word Embeddings Visualization
def visualize_word_embeddings(weights, word_freq, num_words=30, title_suffix=""):
    """
    Visualize word embeddings using t-SNE
    """
    # Get top N words
    top_words = [word for word, _ in word_freq.most_common(num_words)]
    top_indices = [word_to_index[word] for word in top_words]
    top_embeddings = weights[top_indices]
    
    # Apply t-SNE
    tsne = TSNE(n_components=2, random_state=42, perplexity=min(15, num_words-1))
    embeddings_2d = tsne.fit_transform(top_embeddings)
    
    # Create the plot
    plt.figure(figsize=(16, 12))
    scatter = plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1], alpha=0.7, s=100, c=range(num_words), cmap='viridis')
    
    # Add word labels
    for i, word in enumerate(top_words):
        plt.annotate(word, (embeddings_2d[i, 0], embeddings_2d[i, 1]), 
                    fontsize=11, alpha=0.9,
                    bbox=dict(boxstyle="round,pad=0.3", fc="lightyellow", alpha=0.7))
    
    plt.colorbar(scatter, label='Word Rank')
    plt.title(f'Word Embeddings Visualization - Top {num_words} Words {title_suffix}', 
              fontsize=16, fontweight='bold')
    plt.xlabel('t-SNE Component 1')
    plt.ylabel('t-SNE Component 2')
    plt.grid(True, alpha=0.3)
    plt.show()

print(f"\nğŸ¨ GENERATING WORD EMBEDDINGS VISUALIZATION...")
visualize_word_embeddings(weights, word_freq, 30, f"({document_name})")

# 4. Similar Words Analysis
def find_similar_words(word, weights, word_to_index, index_to_word, top_n=10):
    """
    Find similar words based on cosine similarity
    """
    if word not in word_to_index:
        return f"Word '{word}' not in vocabulary"
    
    # Get word vector
    word_idx = word_to_index[word]
    word_vec = weights[word_idx]
    
    # Calculate cosine similarities
    similarities = {}
    for idx, vec in enumerate(weights):
        if idx != word_idx:
            cosine_sim = np.dot(word_vec, vec) / (np.linalg.norm(word_vec) * np.linalg.norm(vec))
            similarities[index_to_word[idx]] = cosine_sim
    
    # Get top N similar words
    similar_words = sorted(similarities.items(), key=lambda x: x[1], reverse=True)[:top_n]
    
    return similar_words

print(f"\nğŸ” SIMILAR WORDS ANALYSIS:")
test_words = ['learning', 'neural', 'network', 'data', 'model', 'training']
for word in test_words:
    if word in word_to_index:
        similar = find_similar_words(word, weights, word_to_index, index_to_word, 8)
        print(f"\nWords similar to '{word}':")
        for similar_word, score in similar:
            print(f"  {similar_word}: {score:.4f}")
    else:
        print(f"âŒ Word '{word}' not in vocabulary")

# 5. Word Prediction Function - FIXED VERSION
def predict_target_word(context_words, model, word_to_index, index_to_word, window_size, top_k=5):
    """
    Predict target word given context words - FIXED to handle window size properly
    """
    # Check if we have the correct number of context words
    expected_words = window_size * 2
    if len(context_words) != expected_words:
        return f"Need exactly {expected_words} context words, but got {len(context_words)}"
    
    # Convert context words to indices
    context_indices = []
    for word in context_words:
        if word in word_to_index:
            context_indices.append(word_to_index[word])
        else:
            return f"Word '{word}' not in vocabulary"
    
    # Reshape for prediction
    context_array = np.array([context_indices])
    
    # Make prediction
    predictions = model.predict(context_array, verbose=0)[0]
    
    # Get top K predictions
    top_indices = np.argsort(predictions)[-top_k:][::-1]
    top_words = [(index_to_word[idx], predictions[idx]) for idx in top_indices]
    
    return top_words

print(f"\nğŸ¯ WORD PREDICTION EXAMPLES:")
# Create prediction examples based on the actual window size
num_context_words = window_size * 2

if num_context_words == 4:  # window_size = 2
    prediction_examples = [
        ['artificial', 'intelligence', 'machine', 'learning'],
        ['neural', 'networks', 'deep', 'learning'],
        ['data', 'processing', 'feature', 'engineering'],
        ['model', 'training', 'validation', 'accuracy']
    ]
elif num_context_words == 8:  # window_size = 4
    prediction_examples = [
        ['artificial', 'intelligence', 'machine', 'learning', 'fundamentals', 'artificial', 'intelligence', 'simulation'],
        ['neural', 'networks', 'deep', 'learning', 'architectures', 'convolutional', 'recurrent', 'transformers'],
        ['data', 'processing', 'feature', 'engineering', 'model', 'training', 'validation', 'testing'],
        ['supervised', 'unsupervised', 'learning', 'algorithms', 'neural', 'networks', 'deep', 'architectures']
    ]
elif num_context_words == 16:  # window_size = 8
    prediction_examples = [
        ['artificial', 'intelligence', 'machine', 'learning', 'fundamentals', 'artificial', 'intelligence', 'simulation', 'human', 'intelligence', 'processes', 'machines', 'processes', 'include', 'learning', 'reasoning'],
        ['neural', 'networks', 'deep', 'learning', 'architectures', 'convolutional', 'recurrent', 'transformers', 'attention', 'mechanisms', 'backpropagation', 'gradient', 'descent', 'optimization', 'algorithms', 'training'],
        ['data', 'processing', 'feature', 'engineering', 'model', 'training', 'validation', 'testing', 'accuracy', 'precision', 'recall', 'metrics', 'evaluation', 'cross', 'validation', 'performance'],
        ['supervised', 'unsupervised', 'learning', 'algorithms', 'neural', 'networks', 'deep', 'architectures', 'convolutional', 'recurrent', 'transformers', 'attention', 'mechanisms', 'backpropagation', 'gradient', 'descent']
    ]
else:
    # Generic examples for any window size
    prediction_examples = []
    # Use the most common words to create context
    common_words = [word for word, _ in word_freq.most_common(num_context_words * 2)]
    for i in range(4):
        start_idx = i * num_context_words
        context = common_words[start_idx:start_idx + num_context_words]
        if len(context) == num_context_words:
            prediction_examples.append(context)

for i, context in enumerate(prediction_examples):
    if len(context) == num_context_words and all(word in word_to_index for word in context):
        predictions = predict_target_word(context, cbow_model, word_to_index, index_to_word, window_size, 5)
        print(f"\nContext {i+1}: {context}")
        print("Top predictions:")
        if isinstance(predictions, list):
            for word, prob in predictions:
                print(f"  {word}: {prob:.4f}")
        else:
            print(f"  {predictions}")
    else:
        print(f"\nContext {i+1}: {context}")
        print("  âŒ Cannot make prediction - context words missing from vocabulary")

# 6. Save Results
def save_all_results(weights, word_to_index, word_freq, history, document_name, window_size):
    """
    Save all results to files
    """
    base_name = document_name.replace(' ', '_').lower()
    
    # Save word embeddings
    embedding_file = f"{base_name}_embeddings.txt"
    with open(embedding_file, 'w', encoding='utf-8') as f:
        f.write(f"{vocab_size} {weights.shape[1]}\n")
        for word, idx in word_to_index.items():
            vec = ' '.join(map(str, weights[idx]))
            f.write(f"{word} {vec}\n")
    
    # Save vocabulary with frequencies
    vocab_file = f"{base_name}_vocabulary.txt"
    with open(vocab_file, 'w', encoding='utf-8') as f:
        f.write("Word\tFrequency\n")
        f.write("----------------\n")
        for word, freq in word_freq.most_common():
            f.write(f"{word}\t{freq}\n")
    
    # Save training history
    history_file = f"{base_name}_training_history.txt"
    with open(history_file, 'w', encoding='utf-8') as f:
        f.write("Training History\n")
        f.write("================\n")
        f.write(f"Window Size: {window_size}\n")
        f.write(f"Final Training Loss: {history.history['loss'][-1]:.4f}\n")
        f.write(f"Final Validation Loss: {history.history['val_loss'][-1]:.4f}\n")
        f.write(f"Final Training Accuracy: {history.history['accuracy'][-1]:.4f}\n")
        f.write(f"Final Validation Accuracy: {history.history['val_accuracy'][-1]:.4f}\n")
    
    print(f"\nğŸ’¾ RESULTS SAVED:")
    print(f"   - Word embeddings: {embedding_file}")
    print(f"   - Vocabulary: {vocab_file}")
    print(f"   - Training history: {history_file}")

save_all_results(weights, word_to_index, word_freq, history, document_name, window_size)

# 7. Final Comprehensive Report
print(f"\nğŸ“ˆ COMPREHENSIVE MODEL EVALUATION REPORT")
print("=" * 50)
print(f"Document: {document_name}")
print(f"Total tokens processed: {len(tokens):,}")
print(f"Vocabulary size: {vocab_size}")
print(f"Training samples generated: {len(X):,}")
print(f"Embedding dimension: {embedding_dim}")
print(f"Context window size: {window_size}")

final_loss, final_accuracy = cbow_model.evaluate(X, y, verbose=0)
print(f"\nğŸ“Š FINAL MODEL PERFORMANCE:")
print(f"Training Loss: {final_loss:.4f}")
print(f"Training Accuracy: {final_accuracy:.4f}")
print(f"Validation Accuracy: {history.history['val_accuracy'][-1]:.4f}")

print(f"\nğŸ”— SEMANTIC RELATIONSHIPS DISCOVERED:")
relationship_pairs = [
    ('learning', 'machine'),
    ('neural', 'network'),
    ('deep', 'learning'),
    ('data', 'training'),
    ('model', 'accuracy')
]

for word1, word2 in relationship_pairs:
    if word1 in word_to_index and word2 in word_to_index:
        vec1 = weights[word_to_index[word1]]
        vec2 = weights[word_to_index[word2]]
        similarity = np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))
        print(f"  {word1} â†” {word2}: {similarity:.4f}")

print(f"\nğŸ‰ CBOW MODEL TRAINING COMPLETE!")
print("=" * 50)
print(f"âœ… Word embeddings successfully learned")
print(f"âœ… Semantic relationships captured")
print(f"âœ… Model ready for word prediction tasks")
print(f"âœ… All results saved to files")
```

```python
# ============================================================
# Continuous Bag of Words (CBOW) Model Implementation for CBOW.txt Final
# ============================================================

# a. Import required libraries
import tensorflow as tf
from tensorflow.keras import models, layers, optimizers
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import re
from sklearn.manifold import TSNE
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

# Download NLTK data (run once)
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')

try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('stopwords')

print(f"TensorFlow version: {tf.__version__}")

# For reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# ============================================================
# a. Data Preparation
# ============================================================

print("=" * 60)
print("STAGE A: DATA PREPARATION")
print("=" * 60)

def load_and_preprocess_text_data():
    """
    Load and preprocess textual data from CBOW.txt file
    """
    try:
        # Load from CBOW.txt file
        with open('CBOW.txt', 'r', encoding='utf-8') as file:
            sample_text = file.read()
        
        print("ğŸ“„ LOADED TEXT FROM CBOW.txt:")
        print(sample_text[:500] + "..." if len(sample_text) > 500 else sample_text)
        
        return sample_text
    except FileNotFoundError:
        print("âŒ CBOW.txt file not found. Using sample text instead.")
        # Fallback sample text
        sample_text = """
        The speed of transmission is an important point of difference between the two viruses. 
        Influenza has a shorter median incubation period and a shorter serial interval than COVID-19 virus. 
        The serial interval for COVID-19 virus is estimated to be 5-6 days, while for influenza virus, the serial interval is 3 days. 
        This means that influenza can spread faster than COVID-19.
        """
        return sample_text

def preprocess_text(text):
    """
    Preprocess the text: lowercase, remove punctuation, tokenize
    """
    # Convert to lowercase
    text = text.lower()
    
    # Remove punctuation and numbers
    text = re.sub(r'[^\w\s]', ' ', text)
    text = re.sub(r'\d+', '', text)
    
    # Tokenize
    tokens = word_tokenize(text)
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words and len(token) > 1]
    
    return tokens

# Load and preprocess data
text_data = load_and_preprocess_text_data()
tokens = preprocess_text(text_data)

print(f"\nğŸ“Š TEXT STATISTICS:")
print(f"Total tokens: {len(tokens)}")
print(f"Unique tokens: {len(set(tokens))}")
print(f"First 20 tokens: {tokens[:20]}")

# Build vocabulary
word_freq = Counter(tokens)
vocab = sorted(word_freq, key=word_freq.get, reverse=True)
word_to_index = {word: idx for idx, word in enumerate(vocab)}
index_to_word = {idx: word for idx, word in enumerate(vocab)}

vocab_size = len(vocab)
print(f"\nğŸ“š VOCABULARY:")
print(f"Vocabulary size: {vocab_size}")
print(f"Most common words: {word_freq.most_common(10)}")

# ============================================================
# b. Generate Training Data
# ============================================================

print("\n" + "=" * 60)
print("STAGE B: GENERATE TRAINING DATA")
print("=" * 60)

def generate_cbow_data(tokens, word_to_index, window_size=2):
    """
    Generate training data for CBOW model
    """
    X = []  # Context words
    y = []  # Target words
    
    for i in range(window_size, len(tokens) - window_size):
        # Get context words (surrounding words)
        context = []
        for j in range(-window_size, window_size + 1):
            if j != 0:  # Skip the target word itself
                context.append(tokens[i + j])
        
        # Get target word
        target = tokens[i]
        
        # Convert to indices
        context_indices = [word_to_index[word] for word in context]
        target_index = word_to_index[target]
        
        X.append(context_indices)
        y.append(target_index)
    
    return np.array(X), np.array(y)

# Generate training data
window_size = 2
X, y = generate_cbow_data(tokens, word_to_index, window_size)

print(f"ğŸ“Š TRAINING DATA SHAPE:")
print(f"Context words (X): {X.shape}")
print(f"Target words (y): {y.shape}")
print(f"Window size: {window_size}")
print(f"Context words per sample: {X.shape[1]}")

# Display sample training data
print(f"\nğŸ” SAMPLE TRAINING DATA:")
for i in range(5):
    context_words = [index_to_word[idx] for idx in X[i]]
    target_word = index_to_word[y[i]]
    print(f"Context: {context_words} -> Target: {target_word}")

# ============================================================
# c. Train Model
# ============================================================

print("\n" + "=" * 60)
print("STAGE C: TRAIN CBOW MODEL")
print("=" * 60)

def build_cbow_model(vocab_size, embedding_dim=50, window_size=2):
    """
    Build CBOW model architecture
    """
    model = models.Sequential([
        # Embedding layer for context words
        layers.Embedding(input_dim=vocab_size, 
                        output_dim=embedding_dim, 
                        input_length=window_size * 2,
                        name='embedding'),
        
        # Average the embeddings of context words
        layers.GlobalAveragePooling1D(name='average_pooling'),
        
        # Dense layer with ReLU activation
        layers.Dense(128, activation='relu', name='hidden_layer'),
        layers.Dropout(0.3),
        
        # Output layer with softmax for word prediction
        layers.Dense(vocab_size, activation='softmax', name='output')
    ])
    
    return model

# Model parameters
embedding_dim = 50
window_size = 2

# Build model
cbow_model = build_cbow_model(vocab_size, embedding_dim, window_size)

print("ğŸ”§ CBOW MODEL ARCHITECTURE:")
cbow_model.summary()

# Compile the model
cbow_model.compile(
    optimizer=optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("\nâœ… MODEL COMPILED:")
print(f"Optimizer: Adam (lr=0.001)")
print(f"Loss: Sparse Categorical Crossentropy")
print(f"Metrics: Accuracy")

# Train the model
print("\nğŸš€ TRAINING CBOW MODEL...")
history = cbow_model.fit(
    X, y,
    epochs=100,
    batch_size=32,
    validation_split=0.2,
    verbose=1,
    shuffle=True
)

print("\nâœ… Training completed!")

# ============================================================
# d. Output
# ============================================================

print("\n" + "=" * 60)
print("STAGE D: OUTPUT AND EVALUATION")
print("=" * 60)

# Training history visualization
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss', linewidth=2)
plt.plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
plt.title('CBOW Model Training Loss', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy', linewidth=2)
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', linewidth=2)
plt.title('CBOW Model Training Accuracy', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Extract word embeddings
embedding_layer = cbow_model.get_layer('embedding')
weights = embedding_layer.get_weights()[0]
print(f"\nğŸ“Š WORD EMBEDDINGS:")
print(f"Embedding matrix shape: {weights.shape}")

# Visualize word embeddings using t-SNE
def visualize_embeddings(weights, words, num_words=20):
    """
    Visualize word embeddings using t-SNE
    """
    # Get top N words
    top_words = words[:num_words]
    top_indices = [word_to_index[word] for word in top_words]
    top_embeddings = weights[top_indices]
    
    # Apply t-SNE
    tsne = TSNE(n_components=2, random_state=42, perplexity=min(5, num_words-1))
    embeddings_2d = tsne.fit_transform(top_embeddings)
    
    # Plot
    plt.figure(figsize=(12, 8))
    plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1], alpha=0.7, s=100)
    
    for i, word in enumerate(top_words):
        plt.annotate(word, (embeddings_2d[i, 0], embeddings_2d[i, 1]), 
                    fontsize=12, alpha=0.8,
                    bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue", alpha=0.7))
    
    plt.title('Word Embeddings Visualization (t-SNE) - CBOW.txt', fontsize=16, fontweight='bold')
    plt.xlabel('t-SNE Component 1')
    plt.ylabel('t-SNE Component 2')
    plt.grid(True, alpha=0.3)
    plt.show()

# Visualize top words
top_words = [word for word, _ in word_freq.most_common(20)]
visualize_embeddings(weights, top_words)

# Function to find similar words
def find_similar_words(word, weights, word_to_index, index_to_word, top_n=5):
    """
    Find similar words based on cosine similarity
    """
    if word not in word_to_index:
        return f"Word '{word}' not in vocabulary"
    
    # Get word vector
    word_idx = word_to_index[word]
    word_vec = weights[word_idx]
    
    # Calculate cosine similarities
    similarities = {}
    for idx, vec in enumerate(weights):
        if idx != word_idx:
            cosine_sim = np.dot(word_vec, vec) / (np.linalg.norm(word_vec) * np.linalg.norm(vec))
            similarities[index_to_word[idx]] = cosine_sim
    
    # Get top N similar words
    similar_words = sorted(similarities.items(), key=lambda x: x[1], reverse=True)[:top_n]
    
    return similar_words

# Test similar words
print("\nğŸ” SIMILAR WORD ANALYSIS:")
test_words = ['virus', 'influenza', 'transmission', 'covid']
for word in test_words:
    if word in word_to_index:
        similar = find_similar_words(word, weights, word_to_index, index_to_word)
        print(f"Words similar to '{word}': {similar}")

# Word prediction function
def predict_target_word(context_words, model, word_to_index, index_to_word, top_k=3):
    """
    Predict target word given context words
    """
    # Convert context words to indices
    context_indices = [word_to_index[word] for word in context_words if word in word_to_index]
    
    if len(context_indices) != window_size * 2:
        return "Invalid context words"
    
    # Reshape for prediction
    context_array = np.array([context_indices])
    
    # Make prediction
    predictions = model.predict(context_array, verbose=0)[0]
    
    # Get top K predictions
    top_indices = np.argsort(predictions)[-top_k:][::-1]
    top_words = [(index_to_word[idx], predictions[idx]) for idx in top_indices]
    
    return top_words

# Test word prediction
print(f"\nğŸ¯ WORD PREDICTION EXAMPLES:")
test_contexts = [
    ['influenza', 'virus', 'serial', 'interval'],
    ['covid', 'virus', 'transmission', 'speed'],
    ['incubation', 'period', 'shorter', 'median']
]

for context in test_contexts:
    predictions = predict_target_word(context, cbow_model, word_to_index, index_to_word)
    print(f"Context: {context} -> Predictions: {predictions}")

# Save word embeddings
def save_embeddings(weights, word_to_index, filename='cbow_embeddings.txt'):
    """
    Save word embeddings to file
    """
    with open(filename, 'w', encoding='utf-8') as f:
        f.write(f"{vocab_size} {embedding_dim}\n")
        for word, idx in word_to_index.items():
            vec = ' '.join(map(str, weights[idx]))
            f.write(f"{word} {vec}\n")
    print(f"âœ… Word embeddings saved to {filename}")

save_embeddings(weights, word_to_index)

# Final model evaluation
print(f"\nğŸ“ˆ FINAL MODEL EVALUATION:")
final_loss, final_accuracy = cbow_model.evaluate(X, y, verbose=0)
print(f"Final Training Loss: {final_loss:.4f}")
print(f"Final Training Accuracy: {final_accuracy:.4f}")

# Display vocabulary statistics
print(f"\nğŸ“š VOCABULARY STATISTICS:")
print(f"Total words in vocabulary: {vocab_size}")
print(f"Embedding dimension: {embedding_dim}")
print(f"Window size: {window_size}")
print(f"Training samples: {len(X)}")

# Show some learned relationships
print(f"\nğŸ”— LEARNED WORD RELATIONSHIPS:")
relationship_examples = [
    (['influenza', 'virus'], 'serial'),
    (['covid', 'virus'], 'transmission'),
    (['incubation', 'period'], 'shorter')
]

for context, expected in relationship_examples:
    if all(word in word_to_index for word in context):
        prediction = predict_target_word(context, cbow_model, word_to_index, index_to_word, top_k=1)
        if prediction and isinstance(prediction, list):
            print(f"Context: {context} | Expected: '{expected}' | Predicted: '{prediction[0][0]}'")

# Additional analysis: Show word frequencies
print(f"\nğŸ“Š WORD FREQUENCY ANALYSIS:")
print("Most frequent words in CBOW.txt:")
for word, freq in word_freq.most_common(15):
    print(f"  {word}: {freq} occurrences")

# Semantic relationships analysis
print(f"\nğŸ” SEMANTIC RELATIONSHIP ANALYSIS:")
medical_terms = ['virus', 'transmission', 'influenza', 'covid', 'symptoms']
for term in medical_terms:
    if term in word_to_index:
        similar = find_similar_words(term, weights, word_to_index, index_to_word, top_n=3)
        print(f"'{term}' relates to: {similar}")

print("\n" + "=" * 70)
print("CBOW MODEL IMPLEMENTATION FOR CBOW.txt COMPLETE!")
print("=" * 70)
```
